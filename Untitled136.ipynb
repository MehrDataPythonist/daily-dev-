```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# داده‌ها
treatment = np.array([94, 197, 16, 38, 99, 141, 23])
control = np.array([52, 104, 146, 10, 50, 31, 40, 27, 46])

# بوت‌استرپ برای فاصله اطمینان نسبت واریانس‌ها
n_bootstraps = 10000
bootstrap_ratios = []
for _ in range(n_bootstraps):
    sample_treatment = np.random.choice(treatment, size=treatment.size, replace=True)
    sample_control = np.random.choice(control, size=control.size, replace=True)
    var_ratio = np.var(sample_treatment, ddof=1) / np.var(sample_control, ddof=1)
    bootstrap_ratios.append(var_ratio)

ci_var_ratio = np.percentile(bootstrap_ratios, [2.5, 97.5])

# بوت‌استرپ برای فاصله اطمینان تفاوت میانگین‌ها
bootstrap_diffs = []
for _ in range(n_bootstraps):
    sample_treatment = np.random.choice(treatment, size=treatment.size, replace=True)
    sample_control = np.random.choice(control, size=control.size, replace=True)
    mean_diff = np.mean(sample_treatment) - np.mean(sample_control)
    bootstrap_diffs.append(mean_diff)

ci_mean_diff = np.percentile(bootstrap_diffs, [2.5, 97.5])

# آزمون فرض
levene_stat, levene_p = stats.levene(treatment, control)
t_stat, t_p = stats.ttest_ind(treatment, control, alternative='greater')

# آزمون‌های ناپارامتری
mannwhitney_stat, mannwhitney_p = stats.mannwhitneyu(treatment, control, alternative='greater')

# بررسی نرمال بودن داده‌ها
fig, ax = plt.subplots(1, 2, figsize=(12, 6))
sns.histplot(treatment, kde=True, ax=ax[0])
ax[0].set_title('توزیع تیمار')
sns.histplot(control, kde=True, ax=ax[1])
ax[1].set_title('توزیع کنترل')
plt.show()

# Q-Q plots
fig, ax = plt.subplots(1, 2, figsize=(12, 6))
stats.probplot(treatment, dist="norm", plot=ax[0])
ax[0].set_title('Q-Q plot تیمار')
stats.probplot(control, dist="norm", plot=ax[1])
ax[1].set_title('Q-Q plot کنترل')
plt.show()

# آزمون‌های نرمالیت
shapiro_treatment = stats.shapiro(treatment)
shapiro_control = stats.shapiro(control)

print(f"فاصله اطمینان نسبت واریانس‌ها: {ci_var_ratio}")
print(f"فاصله اطمینان تفاوت میانگین‌ها: {ci_mean_diff}")
print(f"آزمون Levene: آماره={levene_stat}, p-value={levene_p}")
print(f"آزمون t: آماره={t_stat}, p-value={t_p}")
print(f"آزمون Mann-Whitney U: آماره={mannwhitney_stat}, p-value={mannwhitney_p}")
print(f"آزمون Shapiro-Wilk برای تیمار: آماره={shapiro_treatment[0]}, p-value={shapiro_treatment[1]}")
print(f"آزمون Shapiro-Wilk برای کنترل: آماره={shapiro_control[0]}, p-value={shapiro_control[1]}")
```
