# سلول 1: ساختمان‌های داده پایه (فاز 1)

```python
class Araye:
    def __init__(self, zarfiat=10):
        self.zarfiat = zarfiat
        self.dadeha = [None] * zarfiat
        self.andaze = 0

    def ezafe_kardan(self, dadeh):
        if self.andaze == self.zarfiat:
            self._afzayesh_zarfiat()
        self.dadeha[self.andaze] = dadeh
        self.andaze += 1

    def hazf_kardan(self, index):
        if 0 <= index < self.andaze:
            for i in range(index, self.andaze - 1):
                self.dadeha[i] = self.dadeha[i + 1]
            self.dadeha[self.andaze - 1] = None
            self.andaze -= 1

    def _afzayesh_zarfiat(self):
        self.zarfiat *= 2
        dadeha_jadid = [None] * self.zarfiat
        for i in range(self.andaze):
            dadeha_jadid[i] = self.dadeha[i]
        self.dadeha = dadeha_jadid

    def __iter__(self):
        for i in range(self.andaze):
            yield self.dadeha[i]

    def __getitem__(self, index):
        if 0 <= index < self.andaze:
            return self.dadeha[index]
        raise IndexError("Index out of range")

class Saf:
    def __init__(self):
        self.dadeha = Araye()

    def ezafe_kardan(self, dadeh):
        self.dadeha.ezafe_kardan(dadeh)

    def hazf_kardan(self):
        if not self.khali_ast():
            dadeh = self.dadeha.dadeha[0]
            self.dadeha.hazf_kardan(0)
            return dadeh
        return None

    def khali_ast(self):
        return self.dadeha.andaze == 0

class Poshte:
    def __init__(self):
        self.dadeha = Araye()

    def ezafe_kardan(self, dadeh):
        self.dadeha.ezafe_kardan(dadeh)

    def hazf_kardan(self):
        if not self.khali_ast():
            dadeh = self.dadeha.dadeha[self.dadeha.andaze - 1]
            self.dadeha.hazf_kardan(self.dadeha.andaze - 1)
            return dadeh
        return None

    def khali_ast(self):
        return self.dadeha.andaze == 0
```

# سلول 2: کلاس‌های اصلی (فاز 1 و 2)

```python
class Pezeshk:
    def __init__(self, nam, nam_khanevadegi, kod_melli, shomare_nezam, shomare_tamas, adres, shahr, takhasos, ramz_obour):
        self.nam = nam
        self.nam_khanevadegi = nam_khanevadegi
        self.kod_melli = kod_melli
        self.shomare_nezam = shomare_nezam
        self.shomare_tamas = shomare_tamas
        self.adres = adres
        self.shahr = shahr
        self.takhasos = takhasos
        self.ramz_obour = ramz_obour
        self.nobat_ha = Saf()
        self.tedad_vizit = 0

class Bimar:
    def __init__(self, nam, nam_khanevadegi, kod_melli, shomare_tamas, ramz_obour, jensiat, shahr, shomare_bime):
        self.nam = nam
        self.nam_khanevadegi = nam_khanevadegi
        self.kod_melli = kod_melli
        self.shomare_tamas = shomare_tamas
        self.ramz_obour = ramz_obour
        self.jensiat = jensiat
        self.shahr = shahr
        self.shomare_bime = shomare_bime
        self.tarikhe_nobat_ha = Araye()

class Nobat:
    def __init__(self, pezeshk, bimar, rooz, saat):
        self.pezeshk = pezeshk
        self.bimar = bimar
        self.rooz = rooz
        self.saat = saat
        self.vaziat = "فعال"  # می‌تواند "فعال"، "لغو شده توسط بیمار" یا "لغو شده توسط پزشک" باشد
```

# سلول 3: کلاس SistemeModiriatePezeshki (فاز 1, 2 و 3)

```python
class SistemeModiriatePezeshki:
    def __init__(self):
        self.pezeshkan = Araye()
        self.bimaran = Araye()
        self.shahr_ha = Araye()
        self.takhasos_ha = Araye()
        self.nobat_ha = Araye()
        self.rooz_feli = 1

    def khandan_file_ha(self):
        self._khandan_shahr_ha()
        self._khandan_pezeshkan()
        self._khandan_bimaran()
        self._khandan_takhasos_ha()

    def _khandan_shahr_ha(self):
        with open('Cities.txt', 'r', encoding='utf-8') as file:
            for line in file:
                self.shahr_ha.ezafe_kardan(line.strip())

    def _khandan_pezeshkan(self):
        with open('Doctors.txt', 'r', encoding='utf-8') as file:
            for line in file:
                data = line.strip().split('-')
                if len(data) >= 9:  # اطمینان از وجود حداقل 9 فیلد
                    pezeshk = Pezeshk(*data[:9])  # استفاده از 9 فیلد اول
                    self.pezeshkan.ezafe_kardan(pezeshk)
                else:
                    print(f"خطا در خواندن اطلاعات پزشک: {line}")

    def _khandan_bimaran(self):
        with open('Patients.txt', 'r', encoding='utf-8') as file:
            for line in file:
                data = line.strip().split('-')
                bimar = Bimar(*data)
                self.bimaran.ezafe_kardan(bimar)

    def _khandan_takhasos_ha(self):
        with open('Specialties.txt', 'r', encoding='utf-8') as file:
            for line in file:
                self.takhasos_ha.ezafe_kardan(line.strip())

    def neveshtan_file_ha(self):
        self._neveshtan_shahr_ha()
        self._neveshtan_pezeshkan()
        self._neveshtan_bimaran()
        self._neveshtan_takhasos_ha()

    def _neveshtan_shahr_ha(self):
        with open('Cities.txt', 'w', encoding='utf-8') as file:
            for shahr in self.shahr_ha:
                file.write(f"{shahr}\n")

    def _neveshtan_pezeshkan(self):
        with open('Doctors.txt', 'w', encoding='utf-8') as file:
            for pezeshk in self.pezeshkan:
                line = f"{pezeshk.nam}-{pezeshk.nam_khanevadegi}-{pezeshk.kod_melli}-{pezeshk.shomare_nezam}-{pezeshk.shomare_tamas}-{pezeshk.adres}-{pezeshk.shahr}-{pezeshk.takhasos}-{pezeshk.ramz_obour}\n"
                file.write(line)

    def _neveshtan_bimaran(self):
        with open('Patients.txt', 'w', encoding='utf-8') as file:
            for bimar in self.bimaran:
                line = f"{bimar.nam}-{bimar.nam_khanevadegi}-{bimar.kod_melli}-{bimar.shomare_tamas}-{bimar.ramz_obour}-{bimar.jensiat}-{bimar.shahr}-{bimar.shomare_bime}\n"
                file.write(line)

    def _neveshtan_takhasos_ha(self):
        with open('Specialties.txt', 'w', encoding='utf-8') as file:
            for takhasos in self.takhasos_ha:
                file.write(f"{takhasos}\n")

    def namayesh_pezeshkan(self):
        for pezeshk in self.pezeshkan:
            print(f"نام: {pezeshk.nam} {pezeshk.nam_khanevadegi}, تخصص: {pezeshk.takhasos}, شهر: {pezeshk.shahr}")

    def jostojou_pezeshk_ba_kod_melli(self, kod_melli):
        for pezeshk in self.pezeshkan:
            if pezeshk.kod_melli == kod_melli:
                return pezeshk
        return None

    def namayesh_pezeshk_ba_takhasos(self, takhasos):
        pezeshkan_filtered = Araye()
        for pezeshk in self.pezeshkan:
            if pezeshk.takhasos == takhasos:
                pezeshkan_filtered.ezafe_kardan(pezeshk)
        return pezeshkan_filtered

    def ezafe_kardan_shahr(self, shahr):
        self.shahr_ha.ezafe_kardan(shahr)

    def ezafe_kardan_takhasos(self, takhasos):
        self.takhasos_ha.ezafe_kardan(takhasos)

    def hazf_bimar(self, kod_melli):
        for i in range(self.bimaran.andaze):
            if self.bimaran[i].kod_melli == kod_melli:
                self.bimaran.hazf_kardan(i)
                return True
        return False

    def hazf_pezeshk(self, shomare_nezam):
        for i in range(self.pezeshkan.andaze):
            if self.pezeshkan[i].shomare_nezam == shomare_nezam:
                self.pezeshkan.hazf_kardan(i)
                return True
        return False

    def namayesh_bimaran(self):
        for bimar in self.bimaran:
            print(f"نام: {bimar.nam} {bimar.nam_khanevadegi}, کد ملی: {bimar.kod_melli}, شهر: {bimar.shahr}")

    def ruz_bad(self):
        self.rooz_feli += 1
        print(f"روز جاری به {self.rooz_feli} تغییر کرد.")
```

# سلول 4: توابع پنل بیمار (فاز 2 و 3)

```python
def panele_bimar(sistem):
    while True:
        print("\n--- پنل بیمار ---")
        print("1. ثبت نام")
        print("2. ورود")
        print("3. بازگشت به منوی اصلی")

        entekhab = input("گزینه مورد نظر را انتخاب کنید: ")

        if entekhab == '1':
            sabt_nam_bimar(sistem)
        elif entekhab == '2':
            bimar = vorud_bimar(sistem)
            if bimar:
                menu_bimar(sistem, bimar)
        elif entekhab == '3':
            break
        else:
            print("انتخاب نامعتبر. لطفا دوباره تلاش کنید.")

def sabt_nam_bimar(sistem):
    nam = input("نام: ")
    nam_khanevadegi = input("نام خانوادگی: ")
    kod_melli = input("کد ملی (10 رقم): ")
    shomare_tamas = input("شماره تماس: ")
    ramz_obour = input("رمز عبور: ")
    jensiat = input("جنسیت (مرد/زن): ")

    print("شهرهای موجود:")
    for i, shahr in enumerate(sistem.shahr_ha):
        print(f"{i+1}. {shahr}")
    shahr_index = int(input("شماره شهر را انتخاب کنید: ")) - 1
    shahr = sistem.shahr_ha[shahr_index]

    shomare_bime = input("شماره بیمه: ")

    bimar = Bimar(nam, nam_khanevadegi, kod_melli, shomare_tamas, ramz_obour, jensiat, shahr, shomare_bime)
    sistem.bimaran.ezafe_kardan(bimar)
    print("بیمار با موفقیت ثبت نام شد.")

def vorud_bimar(sistem):
    shomare_tamas = input("شماره تماس: ")
    ramz_obour = input("رمز عبور: ")

    for bimar in sistem.bimaran:
        if bimar.shomare_tamas == shomare_tamas and bimar.ramz_obour == ramz_obour:
            print(f"خوش آمدید {bimar.nam} {bimar.nam_khanevadegi}")
            return bimar

    print("شماره تماس یا رمز عبور اشتباه است.")
    return None

def menu_bimar(sistem, bimar):
    while True:
        print("\n--- منوی بیمار ---")
        print("1. نوبت‌گیری")
        print("2. مشاهده نوبت‌ها")
        print("3. لغو نوبت")
        print("4. رزرو خودکار نوبت")
        print("5. نمایش لیست مکان")
        print("6. نمایش لیست پزشکان")
        print("7. جستجو پزشک")
        print("8. مشاهده تاریخچه نوبت")
        print("9. نمایش نوبت های لغو شده")
        print("10. بازگشت")

        entekhab = input("گزینه مورد نظر را انتخاب کنید: ")

        if entekhab == '1':
            nobat_giri(sistem, bimar)
        elif entekhab == '2':
            namayesh_nobat_haye_bimar(sistem, bimar)
        elif entekhab == '3':
            lagv_nobat_bimar(sistem, bimar)
        elif entekhab == '4':
            rezerv_khodkar_nobat_bimar(sistem, bimar)
        elif entekhab == '5':
            namayesh_list_makan(sistem)
        elif entekhab == '6':
            namayesh_list_pezeshkan(sistem)
        elif entekhab == '7':
            jostojou_pezeshk(sistem)
        elif entekhab == '8':
            namayesh_tarikhe_nobat(sistem, bimar)
        elif entekhab == '9':
            namayesh_nobat_haye_lagv_shode(sistem, bimar)
        elif entekhab == '10':
            break
        else:
            print("انتخاب نامعتبر. لطفا دوباره تلاش کنید.")

def nobat_giri(sistem, bimar):
    print("\n--- نوبت‌گیری ---")
    shahr = input("شهر مورد نظر: ")
    takhasos = input("تخصص مورد نظر: ")
    pezeshkan = sistem.namayesh_pezeshk_ba_takhasos(takhasos)

    if pezeshkan.andaze == 0:
        print("پزشکی با این مشخصات یافت نشد.")
        return

    print("پزشکان موجود:")
    for i, pezeshk in enumerate(pezeshkan):
        print(f"{i+1}. دکتر {pezeshk.nam} {pezeshk.nam_khanevadegi}")

    entekhab_pezeshk = int(input("شماره پزشک مورد نظر را انتخاب کنید: ")) - 1
    if 0 <= entekhab_pezeshk < pezeshkan.andaze:
        pezeshk = pezeshkan[entekhab_pezeshk]
    else:
        print("شماره نامعتبر.")
        return

    rooz = int(input("روز مورد نظر (عدد): "))

    print("ساعت‌های موجود:")
    print("1- ساعت 8 تا 9")
    print("2- ساعت 9 تا 10")
    # ... ادامه ساعت‌ها

    entekhab_saat = int(input("شماره ساعت مورد نظر را انتخاب کنید: "))
    saat = f"{7+entekhab_saat}:00 - {8+entekhab_saat}:00"

    nobat = Nobat(pezeshk, bimar, rooz, saat)
    sistem.nobat_ha.ezafe_kardan(nobat)
    pezeshk.nobat_ha.ezafe_kardan(nobat)
    bimar.tarikhe_nobat_ha.ezafe_kardan(nobat)
    print("نوبت با موفقیت ثبت شد.")

def namayesh_nobat_haye_bimar(sistem, bimar):
    print("\n--- نوبت‌های شما ---")
    for nobat in bimar.tarikhe_nobat_ha:
        if nobat.vaziat == "فعال":
            print(f"روز: {nobat.rooz}, ساعت: {nobat.saat}, پزشک: دکتر {nobat.pezeshk.nam} {nobat.pezeshk.nam_khanevadegi}")

def lagv_nobat_bimar(sistem, bimar):
    namayesh_nobat_haye_bimar(sistem, bimar)
    shomare_nobat = int(input("شماره نوبتی که می‌خواهید لغو کنید را وارد کنید: ")) - 1

    if 0 <= shomare_nobat < bimar.tarikhe_nobat_ha.andaze:
        nobat = bimar.tarikhe_nobat_ha[shomare_nobat]
        if nobat.rooz > sistem.rooz_feli:
            nobat.vaziat = "لغو شده توسط بیمار"
            print("نوبت با موفقیت لغو شد.")
        else:
            print("نوبت‌هایی که در روز جاری یا گذشته هستند را نمی‌توان لغو کرد.")
    else:
        print("شماره نوبت نامعتبر است.")

def rezerv_khodkar_nobat_bimar(sistem, bimar):
    shahr = input("شهر مورد نظر: ")
    takhasos = input("تخصص مورد نظر: ")

    pezeshkan = sistem.namayesh_pezeshk_ba_takhasos(takhasos)
    if pezeshkan.andaze == 0:
        print("پزشکی با این مشخصات یافت نشد.")
        return

    for rooz in range(sistem.rooz_feli, sistem.rooz_feli + 30):
        for pezeshk in pezeshkan:
            for saat in range(8, 20):
                saat_str = f"{saat}:00 - {saat+1}:00"
                if all(nobat.rooz != rooz or nobat.saat != saat_str or nobat.vaziat != "فعال" for nobat in pezeshk.nobat_ha):
                    nobat = Nobat(pezeshk, bimar, rooz, saat_str)
                    sistem.nobat_ha.ezafe_kardan(nobat)
                    pezeshk.nobat_ha.ezafe_kardan(nobat)
                    bimar.tarikhe_nobat_ha.ezafe_kardan(nobat)
                    print(f"نوبت با موفقیت رزرو شد. پزشک: دکتر {pezeshk.nam} {pezeshk.nam_khanevadegi}, روز: {rooz}, ساعت: {saat_str}")
                    return

    print("متأسفانه نوبت خالی در 30 روز آینده یافت نشد.")

def namayesh_list_makan(sistem):
    print("\n--- لیست مکان‌ها ---")
    for shahr in sistem.shahr_ha:
        print(shahr)

def namayesh_list_pezeshkan(sistem):
    print("\n--- لیست پزشکان ---")
    pezeshkan_sorted = sorted(sistem.pezeshkan, key=lambda p: p.tedad_vizit, reverse=True)
    for pezeshk in pezeshkan_sorted:
        print(f"دکتر {pezeshk.nam} {pezeshk.nam_khanevadegi}, تخصص: {pezeshk.takhasos}, شهر: {pezeshk.shahr}, تعداد ویزیت: {pezeshk.tedad_vizit}")

def jostojou_pezeshk(sistem):
    print("\n--- جستجوی پزشک ---")
    jostojou = input("کد ملی یا نام خانوادگی پزشک را وارد کنید: ")
    for pezeshk in sistem.pezeshkan:
        if pezeshk.kod_melli == jostojou or pezeshk.nam_khanevadegi.lower() == jostojou.lower():
            print(f"دکتر {pezeshk.nam} {pezeshk.nam_khanevadegi}, تخصص: {pezeshk.takhasos}, شهر: {pezeshk.shahr}")
            return
    print("پزشکی با این مشخصات یافت نشد.")

def namayesh_tarikhe_nobat(sistem, bimar):
    print("\n--- تاریخچه نوبت‌ها ---")
    for nobat in bimar.tarikhe_nobat_ha:
        print(f"روز: {nobat.rooz}, ساعت: {nobat.saat}, پزشک: دکتر {nobat.pezeshk.nam} {nobat.pezeshk.nam_khanevadegi}, وضعیت: {nobat.vaziat}")

def namayesh_nobat_haye_lagv_shode(sistem, bimar):
    print("\n--- نوبت‌های لغو شده ---")
    for nobat in bimar.tarikhe_nobat_ha:
        if nobat.vaziat != "فعال":
            print(f"روز: {nobat.rooz}, ساعت: {nobat.saat}, پزشک: دکتر {nobat.pezeshk.nam} {nobat.pezeshk.nam_khanevadegi}, وضعیت: {nobat.vaziat}")
```

# سلول 5: توابع پنل پزشک (فاز 2 و 3)

```python
def panele_pezeshk(sistem):
    while True:
        print("\n--- پنل پزشک ---")
        print("1. ثبت نام")
        print("2. ورود")
        print("3. بازگشت به منوی اصلی")

        entekhab = input("گزینه مورد نظر را انتخاب کنید: ")

        if entekhab == '1':
            sabt_nam_pezeshk(sistem)
        elif entekhab == '2':
            pezeshk = vorud_pezeshk(sistem)
            if pezeshk:
                menu_pezeshk(sistem, pezeshk)
        elif entekhab == '3':
            break
        else:
            print("انتخاب نامعتبر. لطفا دوباره تلاش کنید.")

def sabt_nam_pezeshk(sistem):
    nam = input("نام: ")
    nam_khanevadegi = input("نام خانوادگی: ")
    kod_melli = input("کد ملی (10 رقم): ")
    shomare_nezam = input("شماره نظام پزشکی (6 رقم): ")
    shomare_tamas = input("شماره تماس: ")
    adres = input("آدرس: ")

    print("شهرهای موجود:")
    for i, shahr in enumerate(sistem.shahr_ha):
        print(f"{i+1}. {shahr}")
    shahr_index = int(input("شماره شهر را انتخاب کنید: ")) - 1
    shahr = sistem.shahr_ha[shahr_index]

    print("تخصص‌های موجود:")
    for i, takhasos in enumerate(sistem.takhasos_ha):
        print(f"{i+1}. {takhasos}")
    takhasos_index = int(input("شماره تخصص را انتخاب کنید: ")) - 1
    takhasos = sistem.takhasos_ha[takhasos_index]

    ramz_obour = input("رمز عبور: ")

    pezeshk = Pezeshk(nam, nam_khanevadegi, kod_melli, shomare_nezam, shomare_tamas, adres, shahr, takhasos, ramz_obour)
    sistem.pezeshkan.ezafe_kardan(pezeshk)
    print("پزشک با موفقیت ثبت نام شد.")

def vorud_pezeshk(sistem):
    shomare_tamas = input("شماره تماس: ")
    ramz_obour = input("رمز عبور: ")

    for pezeshk in sistem.pezeshkan:
        if pezeshk.shomare_tamas == shomare_tamas and pezeshk.ramz_obour == ramz_obour:
            print(f"خوش آمدید دکتر {pezeshk.nam} {pezeshk.nam_khanevadegi}")
            return pezeshk

    print("شماره تماس یا رمز عبور اشتباه است.")
    return None

def menu_pezeshk(sistem, pezeshk):
    while True:
        print("\n--- منوی پزشک ---")
        print("1. ثبت نوبت")
        print("2. مشاهده نوبت‌ها")
        print("3. لغو نوبت")
        print("4. مشاهده پرونده بیمار")
        print("5. نمایش تاریخچه نوبت‌ها")
        print("6. نمایش نوبت‌های لغو شده")
        print("7. بازگشت")

        entekhab = input("گزینه مورد نظر را انتخاب کنید: ")

        if entekhab == '1':
            sabt_nobat_pezeshk(sistem, pezeshk)
        elif entekhab == '2':
            namayesh_nobat_haye_pezeshk(sistem, pezeshk)
        elif entekhab == '3':
            lagv_nobat_pezeshk(sistem, pezeshk)
        elif entekhab == '4':
            moshahede_parvande_bimar(sistem, pezeshk)
        elif entekhab == '5':
            namayesh_tarikhe_nobat_ha_pezeshk(sistem, pezeshk)
        elif entekhab == '6':
            namayesh_nobat_haye_lagv_shode_pezeshk(sistem, pezeshk)
        elif entekhab == '7':
            break
        else:
            print("انتخاب نامعتبر. لطفا دوباره تلاش کنید.")

def sabt_nobat_pezeshk(sistem, pezeshk):
    print("\n--- ثبت نوبت پزشک ---")
    rooz = int(input("روز مورد نظر (عدد): "))

    print("ساعت‌های موجود:")
    print("1- ساعت 8 تا 9")
    print("2- ساعت 9 تا 10")
    # ... ادامه ساعت‌ها

    while True:
        entekhab_saat = int(input("شماره ساعت مورد نظر را انتخاب کنید (0 برای پایان): "))
        if entekhab_saat == 0:
            break
        saat = f"{7+entekhab_saat}:00 - {8+entekhab_saat}:00"

        nobat = Nobat(pezeshk, None, rooz, saat)
        sistem.nobat_ha.ezafe_kardan(nobat)
        pezeshk.nobat_ha.ezafe_kardan(nobat)
        print(f"نوبت خالی برای روز {rooz} و ساعت {saat} ثبت شد.")

def namayesh_nobat_haye_pezeshk(sistem, pezeshk):
    print("\n--- نوبت‌های شما ---")
    for nobat in pezeshk.nobat_ha:
        if nobat.vaziat == "فعال":
            if nobat.bimar:
                print(f"روز: {nobat.rooz}, ساعت: {nobat.saat}, بیمار: {nobat.bimar.nam} {nobat.bimar.nam_khanevadegi}")
            else:
                print(f"روز: {nobat.rooz}, ساعت: {nobat.saat}, نوبت خالی")

def lagv_nobat_pezeshk(sistem, pezeshk):
    namayesh_nobat_haye_pezeshk(sistem, pezeshk)
    shomare_nobat = int(input("شماره نوبتی که می‌خواهید لغو کنید را وارد کنید: ")) - 1

    if 0 <= shomare_nobat < pezeshk.nobat_ha.andaze:
        nobat = pezeshk.nobat_ha[shomare_nobat]
        if nobat.rooz > sistem.rooz_feli:
            nobat.vaziat = "لغو شده توسط پزشک"
            print("نوبت با موفقیت لغو شد.")
        else:
            print("نوبت‌هایی که در روز جاری یا گذشته هستند را نمی‌توان لغو کرد.")
    else:
        print("شماره نوبت نامعتبر است.")

def moshahede_parvande_bimar(sistem, pezeshk):
    kod_melli = input("کد ملی بیمار را وارد کنید: ")
    bimar = next((b for b in sistem.bimaran if b.kod_melli == kod_melli), None)
    if bimar:
        print(f"\n--- پرونده بیمار: {bimar.nam} {bimar.nam_khanevadegi} ---")
        print(f"کد ملی: {bimar.kod_melli}")
        print(f"جنسیت: {bimar.jensiat}")
        print(f"شهر: {bimar.shahr}")
        print(f"شماره بیمه: {bimar.shomare_bime}")
        print("\nتاریخچه نوبت‌ها:")
        for nobat in bimar.tarikhe_nobat_ha:
            if nobat.pezeshk.kod_melli == pezeshk.kod_melli:
                print(f"روز: {nobat.rooz}, ساعت: {nobat.saat}, وضعیت: {nobat.vaziat}")
    else:
        print("بیماری با این کد ملی یافت نشد.")

def namayesh_tarikhe_nobat_ha_pezeshk(sistem, pezeshk):
    print("\n--- تاریخچه نوبت‌ها ---")
    for nobat in pezeshk.nobat_ha:
        if nobat.bimar:
            print(f"روز: {nobat.rooz}, ساعت: {nobat.saat}, بیمار: {nobat.bimar.nam} {nobat.bimar.nam_khanevadegi}, وضعیت: {nobat.vaziat}")
        else:
            print(f"روز: {nobat.rooz}, ساعت: {nobat.saat}, نوبت خالی، وضعیت: {nobat.vaziat}")

def namayesh_nobat_haye_lagv_shode_pezeshk(sistem, pezeshk):
    print("\n--- نوبت‌های لغو شده ---")
    for nobat in pezeshk.nobat_ha:
        if nobat.vaziat != "فعال":
            if nobat.bimar:
                print(f"روز: {nobat.rooz}, ساعت: {nobat.saat}, بیمار: {nobat.bimar.nam} {nobat.bimar.nam_khanevadegi}, وضعیت: {nobat.vaziat}")
            else:
                print(f"روز: {nobat.rooz}, ساعت: {nobat.saat}, نوبت خالی، وضعیت: {nobat.vaziat}")
```

# سلول 6: توابع پنل مدیریت (فاز 2 و 3)

```python
def panele_modiriat(sistem):
    while True:
        print("\n--- پنل مدیریت ---")
        print(f"روز جاری: {sistem.rooz_feli}")
        print("1. حذف بیمار")
        print("2. حذف پزشک")
        print("3. نمایش بیماران")
        print("4. نمایش پزشکان")
        print("5. افزودن شهر")
        print("6. افزودن تخصص")
        print("7. رفتن به روز بعد")
        print("8. بازگشت به منوی اصلی")

        entekhab = input("گزینه مورد نظر را انتخاب کنید: ")

        if entekhab == '1':
            hazf_bimar(sistem)
        elif entekhab == '2':
            hazf_pezeshk(sistem)
        elif entekhab == '3':
            sistem.namayesh_bimaran()
        elif entekhab == '4':
            sistem.namayesh_pezeshkan()
        elif entekhab == '5':
            ezafe_kardan_shahr(sistem)
        elif entekhab == '6':
            ezafe_kardan_takhasos(sistem)
        elif entekhab == '7':
            sistem.ruz_bad()
        elif entekhab == '8':
            break
        else:
            print("انتخاب نامعتبر. لطفا دوباره تلاش کنید.")

def hazf_bimar(sistem):
    kod_melli = input("کد ملی بیمار را وارد کنید: ")
    if sistem.hazf_bimar(kod_melli):
        print("بیمار با موفقیت حذف شد.")
    else:
        print("بیمار مورد نظر یافت نشد.")

def hazf_pezeshk(sistem):
    shomare_nezam = input("شماره نظام پزشکی را وارد کنید: ")
    if sistem.hazf_pezeshk(shomare_nezam):
        print("پزشک با موفقیت حذف شد.")
    else:
        print("پزشک مورد نظر یافت نشد.")

def ezafe_kardan_shahr(sistem):
    shahr = input("نام شهر جدید را وارد کنید: ")
    sistem.ezafe_kardan_shahr(shahr)
    print(f"شهر {shahr} با موفقیت اضافه شد.")

def ezafe_kardan_takhasos(sistem):
    takhasos = input("نام تخصص جدید را وارد کنید: ")
    sistem.ezafe_kardan_takhasos(takhasos)
    print(f"تخصص {takhasos} با موفقیت اضافه شد.")
```

```python
def main():
    sistem = SistemeModiriatePezeshki()
    sistem.khandan_file_ha()

    while True:
        print("\n--- سیستم مدیریت پزشکی ---")
        print("1. ورود به پنل بیمار")
        print("2. ورود به پنل پزشک")
        print("3. ورود به پنل مدیریت")
        print("4. خروج")

        entekhab = input("گزینه مورد نظر را انتخاب کنید: ")

        if entekhab == '1':
            panele_bimar(sistem)
        elif entekhab == '2':
            panele_pezeshk(sistem)
        elif entekhab == '3':
            panele_modiriat(sistem)
        elif entekhab == '4':
            sistem.neveshtan_file_ha()
            print("با تشکر از استفاده شما از سیستم مدیریت پزشکی. خداحافظ!")
            break
        else:
            print("انتخاب نامعتبر. لطفا دوباره تلاش کنید.")

if __name__ == "__main__":
    main()
```
