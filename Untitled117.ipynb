```python
import pandas as pd
import numpy as np
import jdatetime
from scipy.stats import kstest, expon
import matplotlib.pyplot as plt
import seaborn as sns

# خواندن داده‌ها از فایل اکسل
file_path = '/content/earthquake.xlsx'
df = pd.read_excel(file_path)

# بررسی داده‌ها
print(df.head())
print(df.info())

# تغییر نام ستون‌ها به فینگلیش
df = df.rename(columns={'ساعت': 'saat', 'بزرگی': 'bozorgi', 'عمق': 'omgh'})

# تبدیل ستون 'saat' به رشته
df['saat'] = df['saat'].astype(str)

# حذف مقادیر نامعتبر یا خالی
df = df.dropna(subset=['روز', 'saat'])
df = df[df['روز'].str.strip() != '']

# تابع تبدیل تاریخ شمسی به میلادی و محاسبه زمان به دقیقه
def shamsi_to_datetime(shamsi_date, time_str):
    try:
        year, month, day = map(int, shamsi_date.split('/'))
        hour, minute, _ = map(int, time_str.split(':'))
        jalali_date = jdatetime.date(year, month, day)
        gregorian_date = jalali_date.togregorian()
        return pd.Timestamp(gregorian_date.year, gregorian_date.month, gregorian_date.day, hour, minute)
    except Exception as e:
        print(f"Error converting date: {e}")
        return pd.NaT

# تبدیل تاریخ‌ها و محاسبه فاصله زمانی
df['datetime'] = df.apply(lambda row: shamsi_to_datetime(row['روز'], row['saat']), axis=1)
df = df.dropna(subset=['datetime'])
df = df.sort_values(by='datetime')
df['diff_minutes'] = df['datetime'].diff().dt.total_seconds() / 60

# اولین داده را NaN قرار می‌دهیم
if len(df) > 0:
    df.loc[df.index[0], 'diff_minutes'] = np.nan

# حذف داده‌های NaN
diff_minutes = df['diff_minutes'].dropna()

# برآورد λ
if len(diff_minutes) > 0:
    lambda_est = 1 / diff_minutes.mean()

    # آزمون کولموگروف-اسمیرنوف
    ks_stat, p_value = kstest(diff_minutes, 'expon', args=(0, 1/lambda_est))

    print(f"Estimated λ: {lambda_est}")
    print(f"KS Statistic: {ks_stat}")
    print(f"P-value: {p_value}")

    # نتیجه آزمون
    if p_value > 0.05:
        print("dadeha az tozi'e nemayi peyravi mikonand.")
    else:
        print("dadeha az tozi'e nemayi peyravi nemikonand.")

    # رسم پلات‌های مختلف
    plt.figure(figsize=(14, 10))

    # پلات توزیع فاصله‌های زمانی
    plt.subplot(2, 2, 1)
    sns.histplot(diff_minutes, kde=True, bins=30, color='blue')
    plt.title('Tozi\'e faselehaye zamani bein zelzelehha')
    plt.xlabel('Faseleh zamani (daghighe)')
    plt.ylabel('Tedad')

    # پلات توزیع نمایی شبیه‌سازی شده
    plt.subplot(2, 2, 2)
    expon_data = np.random.exponential(scale=1/lambda_est, size=1000)
    sns.histplot(expon_data, kde=True, bins=30, color='red')
    plt.title('Tozi\'e nemayi shabihsazi shode')
    plt.xlabel('Faseleh zamani (daghighe)')
    plt.ylabel('Tedad')

    # پلات توزیع بزرگی زلزله‌ها
    plt.subplot(2, 2, 3)
    sns.histplot(df['bozorgi'], kde=True, bins=30, color='green')
    plt.title('Tozi\'e bozorgi zelzelehha')
    plt.xlabel('Bozorgi')
    plt.ylabel('Tedad')

    # پلات زمان وقوع زلزله‌ها
    plt.subplot(2, 2, 4)
    sns.scatterplot(x=df['datetime'], y=df['bozorgi'], hue=df['omgh'], palette='viridis')
    plt.title('Zaman voghou\' zelzelehha')
    plt.xlabel('Zaman')
    plt.ylabel('Bozorgi')

    plt.tight_layout()
    plt.show()
else:
    print("hich dade motabari baraye tajziye va tahlil vojoud nadarad.")
```
