```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
import shap
import tensorflow as tf

from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.impute import SimpleImputer

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.regularizers import l2

import matplotlib.pyplot as plt
plt.rcParams['font.family'] = ''
plt.rcParams['font.sans-serif'] = []

warnings.filterwarnings("ignore")
```

##سلول 2: تعریف توابع کمکی



```python
def read_data(file_name):
    data = pd.read_csv(file_name)
    data = data.fillna(data.mean())
    return data

def add_time_features(df, features):
    for feature in features:
        for lag in range(1, 4):
            df[f"{feature}_lag{lag}"] = df[feature].shift(lag)
        for window in [3, 7]:
            df[f"{feature}_rolling_mean{window}"] = df[feature].rolling(window=window).mean()
            df[f"{feature}_rolling_std{window}"] = df[feature].rolling(window=window).std()
        df[f"{feature}_expanding_mean"] = df[feature].expanding().mean()
        df[f"{feature}_expanding_std"] = df[feature].expanding().std()
    return df.fillna(df.mean())

def prepare_data(clean_df, add_time_features_flag=True):
    features = [f'X{i}' for i in range(1, 30)]
    if add_time_features_flag:
        clean_df = add_time_features(clean_df, features)

    Y = clean_df[['Y1', 'Y2', 'Y3', 'Y4', 'Y5']].values
    Sample = clean_df['Sample'].values

    X = clean_df.drop(['Y1', 'Y2', 'Y3', 'Y4', 'Y5', 'Sample', 'Time'], axis=1)

    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    return X_scaled, Y, Sample, X.columns.tolist()

def divide_data(sample, tr_size=0.7):
    unique_vals = np.unique(sample)
    tr_id, te_id = [], []
    for i in unique_vals:
        indices = np.where(sample == i)[0]
        np.random.shuffle(indices)
        split = int(len(indices) * tr_size)
        tr_id.extend(indices[:split])
        te_id.extend(indices[split:])
    return np.array(tr_id), np.array(te_id)
```

##سلول 3: خواندن و آماده‌سازی داده‌ها



```python
file_name = 'Input-Data1.csv'
data = read_data(file_name)
all_data, labels, sample, params_name = prepare_data(data)
all_data_lstm, labels_lstm, sample_lstm, params_name_lstm = prepare_data(data, add_time_features_flag=False)

print("Shape of all_data:", all_data.shape)
print("Shape of all_data_lstm:", all_data_lstm.shape)
print("Shape of labels:", labels.shape)
print("Shape of labels_lstm:", labels_lstm.shape)
print("Length of sample:", len(sample))
print("Length of sample_lstm:", len(sample_lstm))

tr_id, te_id = divide_data(sample, tr_size=0.7)

print("Length of tr_id:", len(tr_id))
print("Length of te_id:", len(te_id))

x_train, x_test = all_data[tr_id], all_data[te_id]
x_train_lstm, x_test_lstm = all_data_lstm[tr_id], all_data_lstm[te_id]
y_train, y_test = labels[tr_id], labels[te_id]

print("Shape of x_train:", x_train.shape)
print("Shape of x_test:", x_test.shape)
print("Shape of x_train_lstm:", x_train_lstm.shape)
print("Shape of x_test_lstm:", x_test_lstm.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)

x_train_lstm = x_train_lstm.reshape((x_train_lstm.shape[0], 1, x_train_lstm.shape[1]))
x_test_lstm = x_test_lstm.reshape((x_test_lstm.shape[0], 1, x_test_lstm.shape[1]))
```

##سلول 4: پاکسازی و مقیاس‌بندی داده‌ها




```python
print("تعداد کل NaN در x_train_lstm:", np.isnan(x_train_lstm).sum())
print("تعداد NaN در هر ویژگی:")
for i in range(x_train_lstm.shape[2]):
    print(f"ویژگی {i}: {np.isnan(x_train_lstm[:, :, i]).sum()}")

imputer = SimpleImputer(strategy='mean')
x_train_lstm_clean = imputer.fit_transform(x_train_lstm.reshape(-1, x_train_lstm.shape[-1])).reshape(x_train_lstm.shape)
x_test_lstm_clean = imputer.transform(x_test_lstm.reshape(-1, x_test_lstm.shape[-1])).reshape(x_test_lstm.shape)

scaler_x = MinMaxScaler()
x_train_lstm_scaled = scaler_x.fit_transform(x_train_lstm_clean.reshape(-1, x_train_lstm_clean.shape[-1])).reshape(x_train_lstm_clean.shape)
x_test_lstm_scaled = scaler_x.transform(x_test_lstm_clean.reshape(-1, x_test_lstm_clean.shape[-1])).reshape(x_test_lstm_clean.shape)

print("\nپس از پاکسازی و مقیاس‌بندی مجدد:")
print("شکل x_train_lstm_scaled:", x_train_lstm_scaled.shape)
print("حداقل x_train_lstm_scaled:", x_train_lstm_scaled.min())
print("حداکثر x_train_lstm_scaled:", x_train_lstm_scaled.max())
print("تعداد NaN در x_train_lstm_scaled:", np.isnan(x_train_lstm_scaled).sum())

scaler_y = MinMaxScaler()
y_train_scaled = scaler_y.fit_transform(y_train)
y_test_scaled = scaler_y.transform(y_test)
```

##سلول 5: تعریف و آموزش مدل‌ها



```python
# قبل از حلقه اصلی
scalers_y = [MinMaxScaler() for _ in range(5)]
y_train_scaled = np.array([scaler.fit_transform(y_train[:, i].reshape(-1, 1)).flatten() for i, scaler in enumerate(scalers_y)]).T

models_per_target = {name: [] for name in ['RandomForest', 'DecisionTree', 'SVR', 'LSTM']}
results = {name: [] for name in ['RandomForest', 'DecisionTree', 'SVR', 'LSTM']}

for i in range(5):  # برای هر متغیر هدف
    for name in ['RandomForest', 'DecisionTree', 'SVR', 'LSTM']:
        if name == 'RandomForest':
            model = RandomForestRegressor(n_estimators=100, random_state=42)
            model.fit(x_train, y_train[:, i])
            predictions = model.predict(x_test)
        elif name == 'DecisionTree':
            model = DecisionTreeRegressor(max_depth=10)
            model.fit(x_train, y_train[:, i])
            predictions = model.predict(x_test)
        elif name == 'SVR':
            model = SVR(kernel='rbf', C=10.0, epsilon=0.1)
            model.fit(x_train, y_train[:, i])
            predictions = model.predict(x_test)
        else:  # LSTM
            model = Sequential([
                LSTM(32, activation='tanh', input_shape=(x_train_lstm_scaled.shape[1], x_train_lstm_scaled.shape[2]),
                     return_sequences=False, kernel_regularizer=l2(0.01)),
                Dropout(0.2),
                Dense(16, activation='relu', kernel_regularizer=l2(0.01)),
                Dense(1)
            ])
            model.compile(optimizer=Adam(learning_rate=0.001), loss='mse')
            history = model.fit(x_train_lstm_scaled, y_train_scaled[:, i], epochs=50, batch_size=32, verbose=0, validation_split=0.2)
            predictions_scaled = model.predict(x_test_lstm_scaled).flatten()
            predictions = scalers_y[i].inverse_transform(predictions_scaled.reshape(-1, 1)).flatten()

            # نمودار خطا برای LSTM
            plt.figure(figsize=(10,6))
            plt.plot(history.history['loss'], label='Khataye Amoozesh')
            plt.plot(history.history['val_loss'], label='Khataye Etebarsanji')
            plt.title(f'Khataye Model LSTM baraye Y{i+1}')
            plt.ylabel('Khata')
            plt.xlabel('Dore')
            plt.legend()
            plt.show()

        models_per_target[name].append(model)

        print(f"مدل: {name}, هدف: Y{i+1}")
        print("شکل پیش‌بینی‌ها:", predictions.shape)
        print("حداقل پیش‌بینی‌ها:", np.min(predictions))
        print("حداکثر پیش‌بینی‌ها:", np.max(predictions))
        print("تعداد NaN:", np.isnan(predictions).sum())
        print("تعداد Inf:", np.isinf(predictions).sum())
        print("شکل y_test:", y_test[:, i].shape)
        print("حداقل y_test:", np.min(y_test[:, i]))
        print("حداکثر y_test:", np.max(y_test[:, i]))
        print()

        mae = mean_absolute_error(y_test[:, i], predictions)
        mse = mean_squared_error(y_test[:, i], predictions)
        results[name].append({'MAE': mae, 'MSE': mse, 'predictions': predictions})

print("آموزش مدل‌ها به پایان رسید.")
```

##سلول 6: نمودارهای مقایسه‌ای



```python
# رسم نمودار مقایسه‌ای MAE و MSE مدل‌ها برای هر متغیر وابسته
fig, axes = plt.subplots(2, 5, figsize=(20, 10))
fig.suptitle('Moghayese Model-ha baraye har Moteghayer Vabaste')

for i in range(5):
    for j, metric in enumerate(['MAE', 'MSE']):
        axes[j, i].bar(results.keys(), [results[name][i][metric] for name in results.keys()])
        axes[j, i].set_title(f'Y{i+1} - {metric}')
        axes[j, i].set_xlabel('Model')
        if i == 0:
            axes[j, i].set_ylabel(metric)
        axes[j, i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

# نمودار خطی مقایسه‌ای پیش‌بینی‌ها با مقادیر واقعی
fig, axes = plt.subplots(1, 5, figsize=(20, 5))
fig.suptitle('Moghayese Pishbini-ha ba Maghadir Vaghe\'i')

for i in range(5):
    axes[i].plot(y_test[:, i], label='Meghdar Vaghe\'i', alpha=0.7)
    for name in results.keys():
        axes[i].plot(results[name][i]['predictions'], label=f'{name} Pishbini', alpha=0.7)
    axes[i].set_title(f'Y{i+1}')
    axes[i].set_xlabel('Shakhes Nemoone')
    if i == 0:
        axes[i].set_ylabel('Meghdar')
    axes[i].legend()

plt.tight_layout()
plt.show()
```

##سلول 7: محاسبه و نمایش مقادیر SHAP



```python
background_data = shap.sample(x_train, 100)
background_data_lstm = x_train_lstm_scaled[:100]

for name in models_per_target.keys():
    for target_idx in range(5):
        if name == 'RandomForest':
            explainer = shap.TreeExplainer(models_per_target[name][target_idx])
            shap_values = explainer.shap_values(x_test)
        elif name in ['DecisionTree', 'SVR']:
            explainer = shap.KernelExplainer(models_per_target[name][target_idx].predict, background_data)
            shap_values = explainer.shap_values(x_test)
        else:  # LSTM
            def f(X):
                return models_per_target[name][target_idx].predict(X.reshape(-1, 1, X.shape[-1]))
            explainer = shap.KernelExplainer(f, background_data_lstm.reshape(-1, background_data_lstm.shape[-1]))
            shap_values = explainer.shap_values(x_test_lstm_scaled.reshape(-1, x_test_lstm_scaled.shape[-1]))

        plt.figure(figsize=(10, 6))
        if name != 'LSTM':
            shap.summary_plot(shap_values, x_test, plot_type="bar", feature_names=params_name, show=False)
        else:
            shap.summary_plot(shap_values, x_test_lstm_scaled.reshape(-1, x_test_lstm_scaled.shape[-1]),
                              plot_type="bar", feature_names=params_name_lstm, show=False)
        plt.title(f'SHAP Summary for {name} on Y{target_idx+1}')
        plt.tight_layout()
        plt.show()

        mean_shap_values = np.abs(shap_values).mean(axis=0)
        print(f'Mean SHAP values for {name} on Y{target_idx+1}:')
        if name != 'LSTM':
            shap_df = pd.DataFrame({'feature': params_name, 'mean_shap': mean_shap_values})
        else:
            shap_df = pd.DataFrame({'feature': params_name_lstm, 'mean_shap': mean_shap_values})

        shap_df = shap_df.sort_values('mean_shap', ascending=False)
        print(shap_df.to_string(index=False))
        print("\n" + "="*50 + "\n")
```
