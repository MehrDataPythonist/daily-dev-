```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Sample data generation
quarters = pd.date_range(start='2004-01-01', periods=68, freq='Q')
monetary_base = np.linspace(200, 910, num=68)
currency_in_circulation = np.linspace(100, 450, num=68)
demand_deposits = np.linspace(50, 300, num=68)
time_deposits = np.linspace(30, 160, num=68)

# Create a DataFrame
data = pd.DataFrame({
    'Quarter': quarters,
    'Monetary Base': monetary_base,
    'Currency in Circulation': currency_in_circulation,
    'Demand Deposits': demand_deposits,
    'Time Deposits': time_deposits
})

# Plotting
sns.set(style="whitegrid")
plt.figure(figsize=(14, 8))
plt.plot(data['Quarter'], data['Monetary Base'], label='Monetary Base', color='blue')
plt.plot(data['Quarter'], data['Currency in Circulation'], label='Currency in Circulation', color='green')
plt.plot(data['Quarter'], data['Demand Deposits'], label='Demand Deposits', color='red')
plt.plot(data['Quarter'], data['Time Deposits'], label='Time Deposits', color='purple')
plt.title('Quarterly Trends in Monetary Aggregates', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Value (Billion IQD)', fontsize=14)
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import scipy.stats as stats
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix

# تنظیم بذر تصادفی برای بازتولیدپذیری
np.random.seed(42)

# ایجاد داده‌های فرضی
num_samples = 1000
data = {
    'annual_income': np.random.normal(50000, 15000, num_samples),
    'debt_to_income_ratio': np.random.normal(0.3, 0.1, num_samples),
    'credit_score': np.random.normal(700, 50, num_samples),
    'num_loans': np.random.randint(1, 10, num_samples),
    'num_delinquencies': np.random.randint(0, 5, num_samples),
    'employment_years': np.random.randint(1, 30, num_samples),
    'default': np.random.binomial(1, 0.2, num_samples)
}

df = pd.DataFrame(data)

# ذخیره داده‌ها به فایل CSV
df.to_csv('loan_data.csv', index=False)

# محاسبه شاخص‌های توصیفی
desc = df.describe()
print(desc)

# رسم هیستوگرام‌ها
df.hist(bins=20, figsize=(14, 10))
plt.tight_layout()
plt.savefig('histograms_question_4.png')
plt.show()

# رسم نمودار scatter و خط رگرسیون تکی
def plot_scatter_with_regression(df, x, y, question_number):
    sns.lmplot(x=x, y=y, data=df, aspect=1.5, ci=None, line_kws={'color': 'red'})
    plt.title(f'Question {question_number}')
    plt.savefig(f'scatter_{x}_question_{question_number}.png')
    plt.show()

explanatory_vars = ['annual_income', 'debt_to_income_ratio', 'credit_score', 'num_loans', 'num_delinquencies', 'employment_years']

for var in explanatory_vars:
    plot_scatter_with_regression(df, var, 'default', 5)

# آماده‌سازی داده‌ها برای مدل‌سازی
X = df[['annual_income', 'debt_to_income_ratio', 'credit_score', 'num_loans', 'num_delinquencies', 'employment_years']]
X = sm.add_constant(X)  # افزودن ثابت به داده‌ها
y = df['default']

# ساخت مدل رگرسیون لوجستیک
model = sm.Logit(y, X).fit()

# نمایش خلاصه مدل
print(model.summary())

# معادله رگرسیون نهایی
coefficients = model.params
equation = f"logit(p) = {coefficients[0]:.4f} + {coefficients[1]:.4f}*annual_income + {coefficients[2]:.4f}*debt_to_income_ratio + "
equation += f"{coefficients[3]:.4f}*credit_score + {coefficients[4]:.4f}*num_loans + {coefficients[5]:.4f}*num_delinquencies + "
equation += f"{coefficients[6]:.4f}*employment_years"
print("معادله رگرسیون نهایی:", equation)

# محاسبه فاصله اطمینان 95 درصدی برای متغیر اول (annual_income)
conf_int = model.conf_int(alpha=0.05)
annual_income_conf_int = conf_int.loc['annual_income']
print(f"95% Confidence Interval for annual_income: ({annual_income_conf_int[0]:.4f}, {annual_income_conf_int[1]:.4f})")

# محاسبه آماره t و p-value برای متغیر annual_income
t_stat_annual_income = coefficients['annual_income'] / model.bse['annual_income']
p_value_annual_income = (1 - stats.t.cdf(abs(t_stat_annual_income), df=model.df_resid)) * 2

print(f"t-statistic for annual_income: {t_stat_annual_income:.4f}, p-value: {p_value_annual_income:.4f}")

# رسم نمودار توزیع t
x = np.linspace(-4, 4, 1000)
y_t_dist = stats.t.pdf(x, df=model.df_resid)
plt.plot(x, y_t_dist, label='t-distribution')
plt.axvline(t_stat_annual_income, color='red', linestyle='--', label=f't-statistic (annual_income) = {t_stat_annual_income:.4f}')
plt.fill_between(x, 0, y_t_dist, where=(abs(x) > stats.t.ppf(0.975, df=model.df_resid)), color='grey', alpha=0.5, label='Rejection region')
plt.xlabel('t-value')
plt.ylabel('Density')
plt.title('Question 10')
plt.legend()
plt.savefig('t_distribution_question_10.png')
plt.show()

# تبدیل y و y_pred به آرایه‌های نامپای
y_np = y.values
y_pred_prob = model.predict(X)
y_pred = (y_pred_prob > 0.5).astype(int)

# محاسبه معیارهای ارزیابی مدل
accuracy = accuracy_score(y_np, y_pred)
precision = precision_score(y_np, y_pred)
recall = recall_score(y_np, y_pred)
conf_matrix = confusion_matrix(y_np, y_pred)

print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print("Confusion Matrix:")
print(conf_matrix)

# پیش‌بینی مقدار متغیر وابسته (default) در میانگین متغیرهای توضیحی
means = df.mean()
logit_predicted_default = coefficients['const'] + coefficients['annual_income'] * means['annual_income'] + coefficients['debt_to_income_ratio'] * means['debt_to_income_ratio']
logit_predicted_default += coefficients['credit_score'] * means['credit_score'] + coefficients['num_loans'] * means['num_loans']
logit_predicted_default += coefficients['num_delinquencies'] * means['num_delinquencies'] + coefficients['employment_years'] * means['employment_years']

predicted_default_probability = 1 / (1 + np.exp(-logit_predicted_default))
print(f"Predicted default probability at mean values: {predicted_default_probability:.4f}")

# تبدیل واحد اندازه‌گیری credit_score از مقیاس 1 تا 850 به مقیاس 0 تا 1
df['credit_score_scaled'] = df['credit_score'] / 850

# ساخت مدل رگرسیون با متغیر جدید
X_scaled = df[['annual_income', 'debt_to_income_ratio', 'credit_score_scaled', 'num_loans', 'num_delinquencies', 'employment_years']]
X_scaled = sm.add_constant(X_scaled)
model_scaled = sm.Logit(y, X_scaled).fit()

# نمایش خلاصه مدل جدید
print(model_scaled.summary())

# افزودن متغیر annual_income به توان دوم
df['annual_income_squared'] = df['annual_income'] ** 2

# ساخت مدل با متغیر جدید
X_poly = df[['annual_income', 'annual_income_squared', 'debt_to_income_ratio', 'credit_score', 'num_loans', 'num_delinquencies', 'employment_years']]
X_poly = sm.add_constant(X_poly)
model_poly = sm.Logit(y, X_poly).fit()

# نمایش خلاصه مدل جدید
print(model_poly.summary())
```
