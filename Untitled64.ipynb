```python
# objective_function.ipynb

import numpy as np
import math

# Define the objective function to maximize (negative NSE)
def objective_function(variables, month_df, HC_initial):
    LC, LT, LS, DCE, DCF, DCI, HCF, HCI, DTE, DTF, DTI, HTF, HTI, DSE, DSF, HSF = variables

    n_data = len(month_df)
    HC = np.zeros(n_data)
    QCE = np.zeros(n_data)
    QCF = np.zeros(n_data)
    QCI = np.zeros(n_data)
    HT = np.zeros(n_data)
    QTE = np.zeros(n_data)
    QTF = np.zeros(n_data)
    QTI = np.zeros(n_data)
    HS = np.zeros(n_data)
    QSE = np.zeros(n_data)
    QSF = np.zeros(n_data)
    q_total = np.zeros(n_data)

    g = 9.81  # Acceleration due to gravity
    # Set the initial value of HC[0] based on the predefined initial condition (HC_initial)
    HC[0] = HC_initial
    QCE[0] = ((math.pi * DCE ** 2) / 4) * math.sqrt(2 * g * max(HC[0], 0))
    QCF[0] = ((math.pi * DCF ** 2) / 4) * math.sqrt(2 * g * max(0, HC[0] - HCF))
    QCI[0] = ((math.pi * DCI ** 2) / 4) * math.sqrt(2 * g * max(0, HC[0] - HCI))
    HT[0] = QCI[0] * LT + max(HT[0] - (LT * (QTE[0] + QTF[0] + QTI[0])), 0)
    QTE[0] = ((math.pi * DTE ** 2) / 4) * math.sqrt(2 * g * max(HT[0], 0))
    QTF[0] = ((math.pi * DTF ** 2) / 4) * math.sqrt(2 * g * max(0, HT[0] - HTF))
    QTI[0] = ((math.pi * DTI ** 2) / 4) * math.sqrt(2 * g * max(0, HT[0] - HTI))
    HS[0] = QTI[0] * LS + max(HS[0] - (LS * (QSE[0] + QSF[0])), 0)
    QSE[0] = ((math.pi * DSE ** 2) / 4) * math.sqrt(2 * g * max(HS[0], 0))
    QSF[0] = ((math.pi * DSF ** 2) / 4) * math.sqrt(2 * g * max(0, HS[0] - HSF))

    # Use the state of the system fromthe end of the previous month as initial condition
    if month_df.index[0] > 0:  # Check if it's not the first month of the first year
        prev_month_df = df[df.index == month_df.index[0] - 1]
        HC[0] = prev_month_df['HC'].iloc[-1]
        HT[0] = prev_month_df['HT'].iloc[-1]
        HS[0] = prev_month_df['HS'].iloc[-1]

    # Loop over each time step within the month
    for i in range(1, n_data):
        HC[i] = month_df['p'].iloc[i] + max(HC[i-1] - (LC * (QCE[i-1] + QCF[i-1] + QCI[i-1])), 0)
        QCE[i] = ((math.pi * DCE ** 2) / 4) * math.sqrt(2 * g * max(HC[i], 0))
        QCF[i] = ((math.pi * DCF ** 2) / 4) * math.sqrt(2 * g * max(0, HC[i] - HCF))
        QCI[i] = ((math.pi * DCI ** 2) / 4) * math.sqrt(2 * g * max(0, HC[i] - HCI))
        HT[i] = QCI[i] * LT + max(HT[i-1] - (LT * (QTE[i-1] + QTF[i-1] + QTI[i-1])), 0)
        QTE[i] = ((math.pi * DTE ** 2) / 4) * math.sqrt(2 * g * max(HT[i], 0))
        QTF[i] = ((math.pi * DTF ** 2) / 4) * math.sqrt(2 * g * max(0, HT[i] - HTF))
        QTI[i] = ((math.pi * DTI ** 2) / 4) * math.sqrt(2 * g * max(0, HT[i] - HTI))
        HS[i] = QTI[i] * LS + max(HS[i-1] - (LS * (QSE[i-1] + QSF[i-1])), 0)
        QSE[i] = ((math.pi * DSE ** 2) / 4) * math.sqrt(2 * g * max(HS[i], 0))
        QSF[i] = ((math.pi * DSF ** 2) / 4) * math.sqrt(2 * g * max(0, HS[i] - HSF))

    # Calculate simulated q values
    q_total = QCF + QTF + QSF

    # Calculate NSE
    nse = 1 - (np.sum((month_df['q_observed'].values - q_total) ** 2) / np.sum((month_df['q_observed'].values - np.mean(month_df['q_observed'].values)) ** 2))

    # Add penalty terms for negative values
    penalty = 0
    if any(x < 0 for x in variables):
        penalty = 1e6

    return -(nse - penalty)  # Return the negative NSE for minimization
```
