```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# فیلترهای بالاگذر
def ideal_high_pass_filter(size, cutoff):
    rows, cols = size
    crow, ccol = rows // 2 , cols // 2
    mask = np.ones((rows, cols), np.uint8)
    center = (crow, ccol)
    for u in range(rows):
        for v in range(cols):
            if np.sqrt((u - crow)**2 + (v - ccol)**2) <= cutoff:
                mask[u, v] = 0
    return mask

def gaussian_high_pass_filter(size, cutoff):
    rows, cols = size
    crow, ccol = rows // 2 , cols // 2
    mask = np.ones((rows, cols), np.float32)
    for u in range(rows):
        for v in range(cols):
            D_uv = np.sqrt((u - crow)**2 + (v - ccol)**2)
            mask[u, v] = 1 - np.exp(-(D_uv**2) / (2 * (cutoff**2)))
    return mask

def butterworth_high_pass_filter(size, cutoff, n):
    rows, cols = size
    crow, ccol = rows // 2 , cols // 2
    mask = np.ones((rows, cols), np.float32)
    for u in range(rows):
        for v in range(cols):
            D_uv = np.sqrt((u - crow)**2 + (v - ccol)**2)
            mask[u, v] = 1 / (1 + (cutoff / D_uv)**(2 * n))
    return mask

# فیلترهای پایین‌گذر
def ideal_low_pass_filter(size, cutoff):
    rows, cols = size
    crow, ccol = rows // 2 , cols // 2
    mask = np.zeros((rows, cols), np.uint8)
    center = (crow, ccol)
    for u in range(rows):
        for v in range(cols):
            if np.sqrt((u - crow)**2 + (v - ccol)**2) <= cutoff:
                mask[u, v] = 1
    return mask

def gaussian_low_pass_filter(size, cutoff):
    rows, cols = size
    crow, ccol = rows // 2 , cols // 2
    mask = np.zeros((rows, cols), np.float32)
    for u in range(rows):
        for v in range(cols):
            D_uv = np.sqrt((u - crow)**2 + (v - ccol)**2)
            mask[u, v] = np.exp(-(D_uv**2) / (2 * (cutoff**2)))
    return mask

def butterworth_low_pass_filter(size, cutoff, n):
    rows, cols = size
    crow, ccol = rows // 2 , cols // 2
    mask = np.zeros((rows, cols), np.float32)
    for u in range(rows):
        for v in range(cols):
            D_uv = np.sqrt((u - crow)**2 + (v - ccol)**2)
            mask[u, v] = 1 / (1 + (D_uv / cutoff)**(2 * n))
    return mask

# اعمال فیلتر در حوزه فرکانس
def apply_filter(image, mask):
    # تبدیل تصویر به فضای فرکانس
    f = np.fft.fft2(image)
    fshift = np.fft.fftshift(f)

    # اعمال فیلتر
    fshift_filtered = fshift * mask

    # تبدیل معکوس به فضای مکان
    f_ishift = np.fft.ifftshift(fshift_filtered)
    img_back = np.fft.ifft2(f_ishift)
    img_back = np.abs(img_back)

    return img_back, mask, fshift, fshift_filtered

# نمایش تصاویر
def namayesh_tasavir(tasavir, onvanha, andaze=(15, 5)):
    plt.figure(figsize=andaze)
    for i, (img, title) in enumerate(zip(tasavir, onvanha)):
        plt.subplot(1, len(tasavir), i + 1)
        plt.imshow(img, cmap='gray')
        plt.title(title)
        plt.axis('off')
    plt.show()

# بارگذاری تصویر
masire_tasvire_asli = '/content/2024-06-01_14-57-37.jpg'
tasvire_asli = cv2.imread(masire_tasvire_asli, cv2.IMREAD_GRAYSCALE)

# پارامترهای فیلتر
cutoff_1 = 60  # فرکانس قطع برای فیلتر
cutoff_2 = 160  # فرکانس قطع برای فیلتر
n = 2  # مرتبه فیلتر باترورث

# اعمال فیلترهای بالاگذر برای فرکانس قطع 60
filtered_ideal_high_60, mask_ideal_high_60, _, _ = apply_filter(tasvire_asli, ideal_high_pass_filter(tasvire_asli.shape, cutoff_1))
filtered_gaussian_high_60, mask_gaussian_high_60, _, _ = apply_filter(tasvire_asli, gaussian_high_pass_filter(tasvire_asli.shape, cutoff_1))
filtered_butterworth_high_60, mask_butterworth_high_60, _, _ = apply_filter(tasvire_asli, butterworth_high_pass_filter(tasvire_asli.shape, cutoff_1, n))

# اعمال فیلترهای بالاگذر برای فرکانس قطع 160
filtered_ideal_high_160, mask_ideal_high_160, _, _ = apply_filter(tasvire_asli, ideal_high_pass_filter(tasvire_asli.shape, cutoff_2))
filtered_gaussian_high_160, mask_gaussian_high_160, _, _ = apply_filter(tasvire_asli, gaussian_high_pass_filter(tasvire_asli.shape, cutoff_2))
filtered_butterworth_high_160, mask_butterworth_high_160, _, _ = apply_filter(tasvire_asli, butterworth_high_pass_filter(tasvire_asli.shape, cutoff_2, n))

# اعمال فیلترهای پایین‌گذر برای فرکانس قطع 60
filtered_ideal_low_60, mask_ideal_low_60, _, _ = apply_filter(tasvire_asli, ideal_low_pass_filter(tasvire_asli.shape, cutoff_1))
filtered_gaussian_low_60, mask_gaussian_low_60, _, _ = apply_filter(tasvire_asli, gaussian_low_pass_filter(tasvire_asli.shape, cutoff_1))
filtered_butterworth_low_60, mask_butterworth_low_60, _, _ = apply_filter(tasvire_asli, butterworth_low_pass_filter(tasvire_asli.shape, cutoff_1, n))

# اعمال فیلترهای پایین‌گذر برای فرکانس قطع 160
filtered_ideal_low_160, mask_ideal_low_160, _, _ = apply_filter(tasvire_asli, ideal_low_pass_filter(tasvire_asli.shape, cutoff_2))
filtered_gaussian_low_160, mask_gaussian_low_160, _, _ = apply_filter(tasvire_asli, gaussian_low_pass_filter(tasvire_asli.shape, cutoff_2))
filtered_butterworth_low_160, mask_butterworth_low_160, _, _ = apply_filter(tasvire_asli, butterworth_low_pass_filter(tasvire_asli.shape, cutoff_2, n))

# نمایش تصاویر اصلی و فیلتر شده
namayesh_tasavir([tasvire_asli, filtered_ideal_high_60, filtered_gaussian_high_60, filtered_butterworth_high_60],
                 ["Original Image", "Ideal High Pass Filtered Image (cutoff=60)", "Gaussian High Pass Filtered Image (cutoff=60)", "Butterworth High Pass Filtered Image (cutoff=60)"])

namayesh_tasavir([filtered_ideal_high_160, filtered_gaussian_high_160, filtered_butterworth_high_160],
                 ["Ideal High Pass Filtered Image (cutoff=160)", "Gaussian High Pass Filtered Image (cutoff=160)", "Butterworth High Pass Filtered Image (cutoff=160)"])

namayesh_tasavir([filtered_ideal_low_60, filtered_gaussian_low_60, filtered_butterworth_low_60],
                 ["Ideal Low Pass Filtered Image (cutoff=60)", "Gaussian Low Pass Filtered Image (cutoff=60)", "Butterworth Low Pass Filtered Image (cutoff=60)"])

namayesh_tasavir([filtered_ideal_low_160, filtered_gaussian_low_160, filtered_butterworth_low_160],
                 ["Ideal Low Pass Filtered Image (cutoff=160)", "Gaussian Low Pass Filtered Image (cutoff=160)", "Butterworth Low Pass Filtered Image (cutoff=160)"])

# نمایش ماسک فیلترهای بالاگذر برای فرکانس قطع 60 و 160
namayesh_tasavir([mask_ideal_high_60, mask_gaussian_high_60, mask_butterworth_high_60],
                 ["Ideal High Pass Filter Mask (cutoff=60)", "Gaussian High Pass Filter Mask (cutoff=60)", "Butterworth High Pass Filter Mask (cutoff=60)"])

namayesh_tasavir([mask_ideal_high_160, mask_gaussian_high_160, mask_butterworth_high_160],
                 ["Ideal High Pass Filter Mask (cutoff=160)", "Gaussian High Pass Filter Mask (cutoff=160)", "Butterworth High Pass Filter Mask (cutoff=160)"])

# نمایش ماسک فیلترهای پایین‌گذر برای فرکانس قطع 60 و 160
namayesh_tasavir([mask_ideal_low_60, mask_gaussian_low_60, mask_butterworth_low_60],
                 ["Ideal Low Pass Filter Mask (cutoff=60)", "Gaussian Low Pass Filter Mask (cutoff=60)", "Butterworth Low Pass Filter Mask (cutoff=60)"])

namayesh_tasavir([mask_ideal_low_160, mask_gaussian_low_160, mask_butterworth_low_160],
                 ["Ideal Low Pass Filter Mask (cutoff=160)", "Gaussian Low Pass Filter Mask (cutoff=160)", "Butterworth Low Pass Filter Mask (cutoff=160)"])
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, ifft

# پارامترهای سیگنال
fs = 1000  # فرکانس نمونه‌برداری (Hz)
t = np.arange(0, 1.0, 1.0/fs)  # محور زمان
f1, f2 = 50, 200  # فرکانس‌های سیگنال

# ایجاد سیگنال با دو فرکانس مختلف
signal = np.sin(2 * np.pi * f1 * t) + 0.5 * np.sin(2 * np.pi * f2 * t)

# تبدیل فوریه سریع (FFT)
signal_fft = fft(signal)

# محور فرکانسی
freqs = np.fft.fftfreq(len(signal), 1/fs)

# فیلتر تاکید فرکانس بالا
cutoff = 100  # فرکانس قطع (Hz)
high_boost_filter = np.array([1 if abs(f) > cutoff else 0 for f in freqs])

# اعمال فیلتر
filtered_fft = signal_fft * high_boost_filter

# تبدیل معکوس فوریه سریع (IFFT)
filtered_signal = ifft(filtered_fft)

# رسم سیگنال‌ها
plt.figure(figsize=(12, 6))

plt.subplot(3, 1, 1)
plt.plot(t, signal)
plt.title("سیگنال اصلی")
plt.xlabel("زمان (ثانیه)")
plt.ylabel("دامنه")

plt.subplot(3, 1, 2)
plt.plot(freqs, np.abs(signal_fft))
plt.title("حوزه فرکانس سیگنال اصلی")
plt.xlabel("فرکانس (Hz)")
plt.ylabel("دامنه")

plt.subplot(3, 1, 3)
plt.plot(t, filtered_signal.real)
plt.title("سیگنال فیلتر شده")
plt.xlabel("زمان (ثانیه)")
plt.ylabel("دامنه")

plt.tight_layout()
plt.show()
```
