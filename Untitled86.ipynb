تعریف توابع فیلترهای بالاگذر





```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# فیلترهای بالاگذر
def ideal_high_pass_filter(size, cutoff):
    rows, cols = size
    crow, ccol = rows // 2 , cols // 2
    mask = np.ones((rows, cols), np.uint8)
    center = (crow, ccol)
    for u in range(rows):
        for v in range(cols):
            if np.sqrt((u - crow)**2 + (v - ccol)**2) <= cutoff:
                mask[u, v] = 0
    return mask

def gaussian_high_pass_filter(size, cutoff):
    rows, cols = size
    crow, ccol = rows // 2 , cols // 2
    mask = np.ones((rows, cols), np.float32)
    for u in range(rows):
        for v in range(cols):
            D_uv = np.sqrt((u - crow)**2 + (v - ccol)**2)
            mask[u, v] = 1 - np.exp(-(D_uv**2) / (2 * (cutoff**2)))
    return mask

def butterworth_high_pass_filter(size, cutoff, n):
    rows, cols = size
    crow, ccol = rows // 2 , cols // 2
    mask = np.ones((rows, cols), np.float32)
    for u in range(rows):
        for v in range(cols):
            D_uv = np.sqrt((u - crow)**2 + (v - ccol)**2)
            mask[u, v] = 1 / (1 + (cutoff / D_uv)**(2 * n))
    return mask
```

تعریف توابع فیلترهای پایین‌گذر

```python
# فیلترهای پایین‌گذر
def ideal_low_pass_filter(size, cutoff):
    rows, cols = size
    crow, ccol = rows // 2 , cols // 2
    mask = np.zeros((rows, cols), np.uint8)
    center = (crow, ccol)
    for u in range(rows):
        for v in range(cols):
            if np.sqrt((u - crow)**2 + (v - ccol)**2) <= cutoff:
                mask[u, v] = 1
    return mask

def gaussian_low_pass_filter(size, cutoff):
    rows, cols = size
    crow, ccol = rows // 2 , cols // 2
    mask = np.zeros((rows, cols), np.float32)
    for u in range(rows):
        for v in range(cols):
            D_uv = np.sqrt((u - crow)**2 + (v - ccol)**2)
            mask[u, v] = np.exp(-(D_uv**2) / (2 * (cutoff**2)))
    return mask

def butterworth_low_pass_filter(size, cutoff, n):
    rows, cols = size
    crow, ccol = rows // 2 , cols // 2
    mask = np.zeros((rows, cols), np.float32)
    for u in range(rows):
        for v in range(cols):
            D_uv = np.sqrt((u - crow)**2 + (v - ccol)**2)
            mask[u, v] = 1 / (1 + (D_uv / cutoff)**(2 * n))
    return mask
```

تابع اعمال فیلتر و نمایش تصاویر

```python
# اعمال فیلتر در حوزه فرکانس
def apply_filter(image, mask):
    # تبدیل تصویر به فضای فرکانس
    f = np.fft.fft2(image)
    fshift = np.fft.fftshift(f)

    # اعمال فیلتر
    fshift_filtered = fshift * mask

    # تبدیل معکوس به فضای مکان
    f_ishift = np.fft.ifftshift(fshift_filtered)
    img_back = np.fft.ifft2(f_ishift)
    img_back = np.abs(img_back)

    return img_back, mask, fshift, fshift_filtered

# نمایش تصاویر
def namayesh_tasavir(tasavir, onvanha, andaze=(15, 5)):
    plt.figure(figsize=andaze)
    for i, (img, title) in enumerate(zip(tasavir, onvanha)):
        plt.subplot(1, len(tasavir), i + 1)
        plt.imshow(img, cmap='gray')
        plt.title(title)
        plt.axis('off')
    plt.show()
```

بارگذاری تصویر و تعیین پارامترهای فیلتر

```python
# بارگذاری تصویر
masire_tasvire_asli = '/content/2024-06-02_20-53-12.jpg'
tasvire_asli = cv2.imread(masire_tasvire_asli, cv2.IMREAD_GRAYSCALE)

# پارامترهای فیلتر
cutoff_1 = 60  # فرکانس قطع برای فیلتر
cutoff_2 = 160  # فرکانس قطع برای فیلتر
n = 2  # مرتبه فیلتر باترورث
```

 اعمال فیلترهای بالاگذر برای فرکانس قطع 60

```python
# اعمال فیلترهای بالاگذر برای فرکانس قطع 60
filtered_ideal_high_60, mask_ideal_high_60, _, _ = apply_filter(tasvire_asli, ideal_high_pass_filter(tasvire_asli.shape, cutoff_1))
filtered_gaussian_high_60, mask_gaussian_high_60, _, _ = apply_filter(tasvire_asli, gaussian_high_pass_filter(tasvire_asli.shape, cutoff_1))
filtered_butterworth_high_60, mask_butterworth_high_60, _, _ = apply_filter(tasvire_asli, butterworth_high_pass_filter(tasvire_asli.shape, cutoff_1, n))

# نمایش تصاویر فیلتر شده بالاگذر برای فرکانس قطع 60
namayesh_tasavir([tasvire_asli, filtered_ideal_high_60, filtered_gaussian_high_60, filtered_butterworth_high_60],
                 ["Original Image", "Ideal High Pass Filtered Image (cutoff=60)", "Gaussian High Pass Filtered Image (cutoff=60)", "Butterworth High Pass Filtered Image (cutoff=60)"])
```
