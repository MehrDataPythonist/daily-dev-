```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset
from torch.optim import Adam
from torch.nn import MSELoss
from scipy.signal import butter, filtfilt
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt

# Load the Excel file and select the 'Total' sheet
file_path = '/content/VRP (new).xlsx'
df = pd.read_excel(file_path, sheet_name='Total')

# Rename columns for easier access
df.columns = ['Date', 'VRP (MW)']

# Replace NaN values with zero in the 'VRP (MW)' column
df['VRP (MW)'].fillna(0, inplace=True)

# Verify the data
print(df.head())

# Convert 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'])

timestamp_column = 'Date'
float_column = 'VRP (MW)'

# Plotting original data
plt.figure(figsize=(25, 6))  # Adjust size as needed
plt.plot(df[timestamp_column], df[float_column])
plt.xlabel('Timestamp')
plt.ylabel('VRP (MW)')
plt.title('Original Data')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)  # Add grid
plt.tight_layout()  # Adjust layout
plt.show()

# Filter the DataFrame for training and testing periods
train_mask = ((df['Date'] < '2021-05-01') | (df['Date'] >= '2021-09-01'))
test_mask = (df['Date'] >= '2021-05-01') & (df['Date'] <= '2021-08-30')
train_df = df[train_mask]
test_df = df[test_mask]

# Verify the filtered data
print(train_df.head())
print(test_df.head())

# Plotting training data
plt.figure(figsize=(25, 6))  # Adjust size as needed
plt.plot(train_df[timestamp_column], train_df[float_column])
plt.xlabel('Timestamp')
plt.ylabel('VRP (MW)')
plt.title('Train Data')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)  # Add grid
plt.tight_layout()  # Adjust layout
plt.show()
```

```python
# Butterworth Low-Pass Filter function
def butterworth_filter(data_values, cutoff_freq, order):
    nyq_freq = 0.5  # Nyquist frequency is half the sampling frequency
    normal_cutoff = cutoff_freq / nyq_freq
    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    filtered_data = filtfilt(b, a, data_values)
    return filtered_data

# Apply filtering to training and testing data
train_values = train_df['VRP (MW)'].to_numpy()
test_values = test_df['VRP (MW)'].to_numpy()
filtered_data_train1 = butterworth_filter(train_values, cutoff_freq=0.1, order=2)
filtered_data_test1 = butterworth_filter(test_values, cutoff_freq=0.1, order=2)

df2 = train_df.copy()
df2['VRP (MW)'] = filtered_data_train1

# Plotting filtered data before downsampling
plt.figure(figsize=(25, 6))  # Adjust size as needed
plt.plot(df2[timestamp_column], df2[float_column])
plt.xlabel('Timestamp')
plt.ylabel('VRP (MW)')
plt.title('After Filtering, Before Downsampling')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)  # Add grid
plt.tight_layout()  # Adjust layout
plt.show()

# Downsampling (e.g., taking every 10th data point)
downsample_factor = 10
filtered_data_train = filtered_data_train1[::downsample_factor]
filtered_data_test = filtered_data_test1[::downsample_factor]

# Convert filtered data to pandas Series
train_series = pd.Series(filtered_data_train)
test_series = pd.Series(filtered_data_test)

# Apply windowed function to data
class DfToXY:
    def __init__(self, window_size=5):
        self.window_size = window_size

    def transform(self, data_series):
        X, y = [], []
        length = len(data_series)
        for i in range(length - self.window_size * 2):
            x_slice = data_series[i:i + self.window_size]
            y_slice = data_series[i + self.window_size:i + self.window_size * 2]
            if np.all(x_slice == 0):
                continue
            X.append(x_slice)
            y.append(y_slice)
        return np.array(X, dtype=np.float32), np.array(y, dtype=np.float32)

data_processor = DfToXY(window_size=5)
X_train, y_train = data_processor.transform(train_series)
X_test, y_test = data_processor.transform(test_series)

# Normalize data using MinMaxScaler
scaler_X = MinMaxScaler(feature_range=(0, 1))
scaler_y = MinMaxScaler(feature_range=(0, 1))
X_train_scaled = scaler_X.fit_transform(X_train)
X_test_scaled = scaler_X.transform(X_test)
y_train_scaled = scaler_y.fit_transform(y_train)
y_test_scaled = scaler_y.transform(y_test)
```

```python
# Define the LSTM Model
class LSTMModel(nn.Module):
    def __init__(self, input_dim, hidden_dim, num_layers, output_dim):
        super(LSTMModel, self).__init__()
        self.lstm = nn.LSTM(input_dim, hidden_dim, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_dim, output_dim)

    def forward(self, x):
        out, (hn, cn) = self.lstm(x)
        out = self.fc(out[:, -1, :])
        return out

# Training setup
model = LSTMModel(input_dim=1, hidden_dim=64, num_layers=2, output_dim=5)
optimizer = Adam(model.parameters())
criterion = MSELoss()

# Prepare training data for PyTorch LSTM
train_data = torch.tensor(X_train_scaled).unsqueeze(-1)  # Add an extra dimension for LSTM input
train_targets = torch.tensor(y_train_scaled)
train_dataset = TensorDataset(train_data, train_targets)
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)

# Initialize a list to store loss values
epoch_losses = []

# Training loop
num_epochs = 100
for epoch in range(num_epochs):
    batch_losses = []
    for i, (data, target) in enumerate(train_loader):
        optimizer.zero_grad()
        outputs = model(data)
        loss = criterion(outputs, target.view(-1, 5))  # Reshape targets to match output
        if torch.isnan(loss).any():
            print(f"NaN detected in loss at epoch {epoch+1}, batch {i+1}")
            continue
        loss.backward()
        optimizer.step()
        batch_losses.append(loss.item())
    epoch_loss = sum(batch_losses) / len(batch_losses) if batch_losses else float('nan')
    epoch_losses.append(epoch_loss)
    print(f"Epoch {epoch+1}/{num_epochs}, Average Loss: {epoch_loss:.4f}")

print("Training complete!")

# Plotting the training loss
plt.figure(figsize=(10, 5))
plt.plot(epoch_losses, label='Training Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training Loss Over Epochs')
plt.legend()
plt.show()
```

```python
# Making predictions on the test set
model.eval()
with torch.no_grad():
    test_data = torch.tensor(X_test_scaled).unsqueeze(-1)
    predictions = model(test_data).numpy()

actual = scaler_y.inverse_transform(y_test_scaled)
predictions_rescaled = scaler_y.inverse_transform(predictions)

# Plotting model predictions against actual values
plt.figure(figsize=(10, 5))
plt.plot(actual.flatten(), label='Actual Values')
plt.plot(predictions_rescaled.flatten(), label='Predicted Values', linestyle='--')
plt.xlabel('Sample Index')
plt.ylabel('Output Value')
plt.title('Comparison of Predictions and Actual Values')
plt.legend()
plt.show()

# Calculate the average of the five predictions
average_predictions = np.mean(predictions, axis=1)

# We need to duplicate the average_predictions to match the scaler's expected shape
average_predictions_duplicated = np.tile(average_predictions.reshape(-1, 1), (1, predictions.shape[1]))
average_predictions_rescaled = scaler_y.inverse_transform(average_predictions_duplicated)[:, 0]

# Plot the average predicted values
plt.figure(figsize=(10, 5))
plt.plot(average_predictions_rescaled.flatten(), label='Average Predicted Values', linestyle='--')
plt.xlabel('Sample Index')
plt.ylabel('Output Value')
plt.title('Average Predicted Values Over Time')
plt.legend()
plt.show()

# Example: Annotate a specific date (let's assume index 2 for demonstration)
date_index = 2
predicted_average_value = average_predictions_rescaled[date_index]

plt.figure(figsize=(10, 5))
plt.plot(average_predictions_rescaled.flatten(), label='Average Predicted Values', linestyle='--')
plt.xlabel('Sample Index')
plt.ylabel('Output Value')
plt.title('Average Predicted Values Over Time')
plt.legend()

# Annotate the specific date
plt.annotate(f'Predicted Avg on {test_df.iloc[date_index]["Date"].date()}: {predicted_average_value:.2f}',
             xy=(date_index, predicted_average_value), xytext=(date_index+5, predicted_average_value+5),
             arrowprops=dict(facecolor='black', arrowstyle='->'))

plt.show()
```
