```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset
from torch.optim import Adam
from torch.nn import MSELoss
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt

# بارگذاری داده‌ها از فایل اکسل
file_path = '/content/VRP_final.xlsx'
summary_df = pd.read_excel(file_path, sheet_name='Summary')
total_df = pd.read_excel(file_path, sheet_name='Total')

# تبدیل ستون‌های زمانی به فرمت datetime
summary_df['t0_VRP'] = pd.to_datetime(summary_df['t0_VRP'], format='%d/%m/%Y %H:%M', errors='coerce')
total_df['Date (dd/mm/yyyy hh:mm)'] = pd.to_datetime(total_df['Date (dd/mm/yyyy hh:mm)'], format='%Y-%m-%d %H:%M:%S', errors='coerce')

# تبدیل مقادیر خالی ستون VRP به صفر
vrp_column = 'VRP [MW]'
total_df[vrp_column] = total_df[vrp_column].fillna(0)

# حذف ردیف‌هایی که ستون‌های زمانی آنها نامعتبر است
summary_df = summary_df.dropna(subset=['t0_VRP', 'Samples_[t0_vrp,t2_vrp]']).reset_index(drop=True)
```

```python
# اندازه ویندوز
window_size = 5

# آماده‌سازی داده‌ها برای مدل
all_X, all_y = [], []
skipped_events = 0

for i in range(len(summary_df)):
    t0_vrp = summary_df.loc[i, 't0_VRP']
    t2 = summary_df.loc[i, 't2']
    samples_t0_t2 = summary_df.loc[i, 'Samples_[t0_vrp,t2_vrp]']

    # استخراج نمونه‌ها از t0_VRP تا t2
    samples = total_df[(total_df['Date (dd/mm/yyyy hh:mm)'] >= t0_vrp) &
                       (total_df['Date (dd/mm/yyyy hh:mm)'] <= t2)][vrp_column].to_numpy()

    # حذف رویدادهایی که تعداد نمونه‌هایشان کمتر از اندازه ویندوز است
    if len(samples) < window_size:
        skipped_events += 1
        continue

    # ورودی‌ها و خروجی‌ها را تنظیم می‌کنیم
    X = samples[:window_size]
    y = samples_t0_t2 - window_size

    all_X.append(X)
    all_y.append(y)

# تبدیل لیست‌ها به آرایه‌های numpy
all_X = np.array(all_X, dtype=np.float32)
all_y = np.array(all_y, dtype=np.float32)

# نرمال‌سازی داده‌ها با استفاده از MinMaxScaler
scaler_X = MinMaxScaler(feature_range=(0, 1))
scaler_y = MinMaxScaler(feature_range=(0, 1))
X_scaled = scaler_X.fit_transform(all_X)
y_scaled = scaler_y.fit_transform(all_y.reshape(-1, 1))

# تعداد رویدادهایی که استفاده شده‌اند
used_events = len(all_X)
```

```python
# آماده‌سازی داده‌ها برای PyTorch
train_data = torch.tensor(X_scaled).unsqueeze(-1)  # افزودن یک بعد اضافی برای ورودی LSTM
train_targets = torch.tensor(y_scaled)
train_dataset = TensorDataset(train_data, train_targets)
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)

# تعریف مدل LSTM
class LSTMModel(nn.Module):
    def __init__(self, input_dim, hidden_dim, num_layers, output_dim):
        super(LSTMModel, self).__init__()
        self.lstm = nn.LSTM(input_dim, hidden_dim, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_dim, output_dim)

    def forward(self, x):
        out, (hn, cn) = self.lstm(x)
        out = self.fc(out[:, -1, :])
        return out

# تنظیمات آموزش
input_dim = 1
hidden_dim = 64
num_layers = 2
output_dim = 1

model = LSTMModel(input_dim=input_dim, hidden_dim=hidden_dim, num_layers=num_layers, output_dim=output_dim)
optimizer = Adam(model.parameters(), lr=0.001)
criterion = MSELoss()
```

```python
# آماده‌سازی داده‌ها برای PyTorch
train_data = torch.tensor(X_scaled).unsqueeze(-1)  # افزودن یک بعد اضافی برای ورودی LSTM
train_targets = torch.tensor(y_scaled)
train_dataset = TensorDataset(train_data, train_targets)
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)

# تعریف مدل LSTM
class LSTMModel(nn.Module):
    def __init__(self, input_dim, hidden_dim, num_layers, output_dim):
        super(LSTMModel, self).__init__()
        self.lstm = nn.LSTM(input_dim, hidden_dim, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_dim, output_dim)

    def forward(self, x):
        out, (hn, cn) = self.lstm(x)
        out = self.fc(out[:, -1, :])
        return out

# تنظیمات آموزش
input_dim = 1
hidden_dim = 64
num_layers = 2
output_dim = 1

model = LSTMModel(input_dim=input_dim, hidden_dim=hidden_dim, num_layers=num_layers, output_dim=output_dim)
optimizer = Adam(model.parameters(), lr=0.001)
criterion = MSELoss()
```

```python
# حلقه آموزش مدل LSTM
num_epochs = 100
epoch_losses = []

for epoch in range(num_epochs):
    batch_losses = []
    for data, target in train_loader:
        optimizer.zero_grad()
        outputs = model(data)
        loss = criterion(outputs, target)
        loss.backward()
        optimizer.step()
        batch_losses.append(loss.item())
    epoch_loss = sum(batch_losses) / len(batch_losses)
    epoch_losses.append(epoch_loss)
    print(f"Epoch {epoch+1}/{num_epochs}, Average Loss: {epoch_loss:.4f}")

print("Training LSTM complete!")
```

```python
# تعریف مدل MLP
class MLPModel(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim):
        super(MLPModel, self).__init__()
        self.fc1 = nn.Linear(input_dim, hidden_dim)
        self.fc2 = nn.Linear(hidden_dim, output_dim)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = self.fc2(x)
        return x

# تنظیمات آموزش
input_dim = window_size
hidden_dim = 64
output_dim = 1

model_mlp = MLPModel(input_dim=input_dim, hidden_dim=hidden_dim, output_dim=output_dim)
optimizer_mlp = Adam(model_mlp.parameters(), lr=0.001)
criterion_mlp = MSELoss()
```

```python
# آماده‌سازی داده‌ها برای PyTorch (MLP)
train_data_mlp = torch.tensor(X_scaled)
train_targets_mlp = torch.tensor(y_scaled)
train_dataset_mlp = TensorDataset(train_data_mlp, train_targets_mlp)
train_loader_mlp = DataLoader(train_dataset_mlp, batch_size=32, shuffle=True)

# حلقه آموزش مدل MLP
epoch_losses_mlp = []

for epoch in range(num_epochs):
    batch_losses = []
    for data, target in train_loader_mlp:
        optimizer_mlp.zero_grad()
        outputs = model_mlp(data)
        loss = criterion_mlp(outputs, target)
        loss.backward()
        optimizer_mlp.step()
        batch_losses.append(loss.item())
    epoch_loss = sum(batch_losses) / len(batch_losses)
    epoch_losses_mlp.append(epoch_loss)
    print(f"Epoch {epoch+1}/{num_epochs}, Average Loss: {epoch_loss:.4f}")

print("Training MLP complete!")
```

```python
# پیش‌بینی با مدل LSTM
model.eval()
with torch.no_grad():
    predictions_lstm = model(torch.tensor(X_scaled).unsqueeze(-1)).numpy()

# برگرداندن مقیاس داده‌ها
predictions_lstm_rescaled = scaler_y.inverse_transform(predictions_lstm)

# پیش‌بینی با مدل MLP
model_mlp.eval()
with torch.no_grad():
    predictions_mlp = model_mlp(torch.tensor(X_scaled)).numpy()

# برگرداندن مقیاس داده‌ها
predictions_mlp_rescaled = scaler_y.inverse_transform(predictions_mlp)
```

```python
# رسم مقادیر پیش‌بینی شده در مقابل مقادیر واقعی برای مدل LSTM
plt.figure(figsize=(15, 7))
plt.plot(all_y, label='Actual Values', color='blue')
plt.plot(predictions_lstm_rescaled, label='Predicted Values (LSTM)', linestyle='--', color='orange')
plt.plot(predictions_mlp_rescaled, label='Predicted Values (MLP)', linestyle='--', color='green')
plt.xlabel('Sample Index')
plt.ylabel('Output Value')
plt.title('Comparison of Predicted and Actual Values (LSTM & MLP)')
plt.legend()
plt.show()
```

```python
# محاسبه خطای مطلق برای هر رویداد
errors_lstm = np.abs(all_y - predictions_lstm_rescaled.flatten())
errors_mlp = np.abs(all_y - predictions_mlp_rescaled.flatten())

# رسم نمودار خطای مطلق برای هر رویداد
plt.figure(figsize=(15, 7))
plt.plot(errors_lstm, label='Absolute Error (LSTM)', linestyle='-', color='orange')
plt.plot(errors_mlp, label='Absolute Error (MLP)', linestyle='-', color='green')
plt.xlabel('Event Index')
plt.ylabel('Absolute Error')
plt.title('Absolute Error of Predictions for Each Event (LSTM & MLP)')
plt.legend()
plt.show()
```

```python
# رسم نمودار تعداد رویدادهای استفاده شده و اسکیپ شده
plt.figure(figsize=(10, 5))
plt.bar(['Used Events', 'Skipped Events'], [used_events, skipped_events], color=['blue', 'red'])
plt.xlabel('Event Type')
plt.ylabel('Count')
plt.title('Count of Used and Skipped Events Based on Window Size')
plt.show()
```
