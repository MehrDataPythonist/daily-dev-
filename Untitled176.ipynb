```python
import networkx as nx
import matplotlib.pyplot as plt
import random

def calculate_energy(G):
    """
    محاسبه انرژی (هامیلتونی) یک گراف
    """
    energy = 0
    for edge in G.edges():
        i, j = edge
        energy += G.nodes[i]['opinion'] * G.edges[i, j]['weight'] * G.nodes[j]['opinion']
    return energy

def create_graph(n, edge_type):
    """
    ایجاد گراف با نوع یال مشخص
    """
    G = nx.Graph()
    G.add_nodes_from(range(n))

    for i in range(n):
        for j in range(i+1, n):
            if edge_type == 'positive':
                G.add_edge(i, j, weight=1)
            elif edge_type == 'negative':
                G.add_edge(i, j, weight=-1)
            elif edge_type == 'random':
                if random.random() < 0.5:
                    G.add_edge(i, j, weight=random.choice([-1, 1]))

    # تخصیص عقیده به گره‌ها
    for node in G.nodes():
        G.nodes[node]['opinion'] = random.choice([-1, 1])

    return G

def main():
    n = 10  # تعداد رأس‌ها

    graph_types = ['positive', 'negative', 'random']

    for graph_type in graph_types:
        G = create_graph(n, graph_type)
        energy = calculate_energy(G)

        print(f"\nنوع گراف: {graph_type}")
        print(f"انرژی (هامیلتونی) گراف: {energy}")

        # نمایش گراف
        plt.figure(figsize=(6, 6))
        pos = nx.spring_layout(G)
        nx.draw(G, pos, with_labels=True, node_color=['red' if G.nodes[node]['opinion'] == -1 else 'blue' for node in G.nodes()])
        edge_labels = nx.get_edge_attributes(G, 'weight')
        nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
        plt.title(f"گراف {graph_type}")
        plt.show()

if __name__ == "__main__":
    main()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import random

def hamiltonian_energy(G):
    """
    محاسبه انرژی هامیلتونی گراف
    """
    energy = 0
    for edge in G.edges():
        i, j = edge
        energy += G.nodes[i]['opinion'] * G.edges[i, j]['weight'] * G.nodes[j]['opinion']
    return energy

def create_graph(n, type):
    """
    ایجاد گراف با n گره و نوع مشخص
    """
    G = nx.Graph()
    G.add_nodes_from(range(n))

    for i in range(n):
        for j in range(i+1, n):
            if type == 'positive':
                G.add_edge(i, j, weight=1)
            elif type == 'negative':
                G.add_edge(i, j, weight=-1)
            elif type == 'random':
                G.add_edge(i, j, weight=random.choice([-1, 1]))

    # تخصیص عقیده به گره‌ها
    for node in G.nodes():
        G.nodes[node]['opinion'] = random.choice([-1, 1])

    return G

def draw_graph(G, title, energy):
    """
    نمایش گراف با رنگ‌آمیزی گره‌ها و نمایش وزن یال‌ها
    """
    plt.figure(figsize=(6, 6))
    pos = nx.spring_layout(G)

    # رنگ‌آمیزی گره‌ها بر اساس عقیده
    node_colors = ['red' if G.nodes[node]['opinion'] == -1 else 'blue' for node in G.nodes()]

    nx.draw(G, pos, with_labels=True, node_color=node_colors)

    # نمایش وزن یال‌ها
    edge_labels = nx.get_edge_attributes(G, 'weight')
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

    plt.title(title)
    plt.xlabel(f"Energy: {energy:.2f}")
    plt.show()

# ایجاد گراف‌های مختلف
graph_types = ['positive', 'negative', 'random']
n = 10  # تعداد گره‌ها

for graph_type in graph_types:
    G = create_graph(n, graph_type)
    energy = hamiltonian_energy(G)
    draw_graph(G, f"Geraf {graph_type.capitalize()}", energy)

# ایجاد و نمایش گراف‌های مصنوعی
# Erdős-Rényi
G_er = nx.erdos_renyi_graph(n, 0.5)
for (u, v) in G_er.edges():
    G_er.edges[u, v]['weight'] = random.choice([-1, 1])
for node in G_er.nodes():
    G_er.nodes[node]['opinion'] = random.choice([-1, 1])
energy_er = hamiltonian_energy(G_er)
draw_graph(G_er, "Erdős-Rényi Graph", energy_er)

# Watts-Strogatz
G_ws = nx.watts_strogatz_graph(n, 4, 0.3)
for (u, v) in G_ws.edges():
    G_ws.edges[u, v]['weight'] = random.choice([-1, 1])
for node in G_ws.nodes():
    G_ws.nodes[node]['opinion'] = random.choice([-1, 1])
energy_ws = hamiltonian_energy(G_ws)
draw_graph(G_ws, "Watts-Strogatz Graph", energy_ws)

# Barabási-Albert
G_ba = nx.barabasi_albert_graph(n, 2)
for (u, v) in G_ba.edges():
    G_ba.edges[u, v]['weight'] = random.choice([-1, 1])
for node in G_ba.nodes():
    G_ba.nodes[node]['opinion'] = random.choice([-1, 1])
energy_ba = hamiltonian_energy(G_ba)
draw_graph(G_ba, "Barabási-Albert Graph", energy_ba)
```
