```python
value = [1, 3, 4]
A = 7
N = len(value)

def coin_asli(A, N, value):
    B = [[0] * (A + 1) for _ in range(N + 1)]
    for i in range(N + 1):
        for j in range(A + 1):
            if j == 0:
                B[i][j] = 0
            elif i == 1 and j < value[0]:
                B[i][j] = float('inf')
            elif i == 1 and j >= value[0]:
                B[i][j] = 1 + B[i][j - value[0]]
            elif i != 1 and i != 0 and j < value[i - 1]:
                B[i][j] = B[i - 1][j]
            elif i != 1 and i != 0 and j >= value[i - 1]:
                H = 1 + B[i][j - value[i - 1]]
                T = B[i - 1][j]
                if H > T:
                    B[i][j] = T
                else:
                    B[i][j] = H

    return B

B = coin_asli(A, N, value)
cost = B[N][A]
print(f"Minimum number of coins needed: {cost}")

def calculate_coin_count(B, value, N, A):
    selected_coins = [0] * N
    i = N
    j = A

    while i > 0 and j > 0:
        if B[i][j] != B[i - 1][j]:
            selected_coins[i - 1] += 1
            j -= value[i - 1]
        else:
            i -= 1

    return selected_coins

selected_coins = calculate_coin_count(B, value, N, A)

print("Coins used:")

for i in range(len(selected_coins)):
    if selected_coins[i] > 0:
        print(selected_coins[i], "coin(s) of coins", value[i])
```
