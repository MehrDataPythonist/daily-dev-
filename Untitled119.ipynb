```python
import networkx as nx
import matplotlib.pyplot as plt

#    
G = nx.DiGraph()

#   
nodes = ["A", "B", "C", "D", "E", "F", "G"]
G.add_nodes_from(nodes)

#      
edges = [("A", "C"), ("B", "C"), ("C", "E"), ("C", "F"), ("D", "F"), ("E", "G"), ("F", "G")]
G.add_edges_from(edges)

#  
plt.figure(figsize=(10, 6))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=1500, node_color='skyblue', font_size=15, font_weight='bold', arrowsize=20)
nx.draw_networkx_edge_labels(G, pos, edge_labels={(u, v): '' for u, v in edges}, font_color='red')
plt.title('Dependency Graph')
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

#   
G = nx.Graph()

#   
nodes = [1, 2, 3, 4, 5]
G.add_nodes_from(nodes)

#      
edges = [(1, 2), (1, 3), (1, 4), (2, 5), (2, 4), (3, 5), (4, 5)]
G.add_edges_from(edges)

#    
colors = {1: 'green', 2: 'blue', 3: 'red', 4: 'red', 5: 'green'}

#  
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=1500, node_color=[colors[node] for node in G.nodes], font_size=15, font_weight='bold', edge_color='black')
plt.title('Graph Coloring')
plt.show()
```
