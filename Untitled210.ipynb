```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from datetime import datetime

# 1. خواندن داده‌ها
file_path = "/content/ev_data/EVChargingStationUsage.csv"
df = pd.read_csv(file_path)

# 2. پر کردن مقادیر گمشده

# پر کردن مقادیر عددی با میانگین
numeric_columns = ['Energy (kWh)', 'GHG Savings (kg)', 'Gasoline Savings (gallons)', 'Fee', 'Driver Postal Code', 'EVSE ID', 'System S/N']
for col in numeric_columns:
    df[col].fillna(df[col].mean(), inplace=True)

# پر کردن مقادیر دسته‌بندی با مد (بیشترین مقدار تکراری)
categorical_columns = ['Port Type', 'Plug Type', 'Currency', 'Ended By', 'User ID', 'County', 'Model Number']
for col in categorical_columns:
    df[col].fillna(df[col].mode()[0], inplace=True)

# 3. تبدیل ستون‌های زمانی به datetime
df['Start Date'] = pd.to_datetime(df['Start Date'], errors='coerce')
df['End Date'] = pd.to_datetime(df['End Date'], errors='coerce')

# پر کردن مقادیر گمشده‌ی زمانی با میانگین زمانی
df['Start Date'].fillna(df['Start Date'].mean(), inplace=True)
df['End Date'].fillna(df['End Date'].mean(), inplace=True)

# 4. ایجاد ویژگی‌های زمانی
df['Start Hour'] = df['Start Date'].dt.hour
df['Start Day'] = df['Start Date'].dt.day
df['Start Month'] = df['Start Date'].dt.month
df['Start Year'] = df['Start Date'].dt.year
df['Start Weekday'] = df['Start Date'].dt.weekday  # روز هفته (0 = دوشنبه, 6 = یکشنبه)

df['End Hour'] = df['End Date'].dt.hour
df['End Day'] = df['End Date'].dt.day
df['End Month'] = df['End Date'].dt.month
df['End Year'] = df['End Date'].dt.year
df['End Weekday'] = df['End Date'].dt.weekday

# 5. تبدیل داده‌های دسته‌بندی به اعداد (Label Encoding)
label_columns = ['Station Name', 'MAC Address', 'Org Name', 'Port Type', 'Plug Type', 'Currency', 'Ended By', 'County', 'Model Number']
label_encoders = {}
for col in label_columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le

# 6. نرمال‌سازی داده‌های عددی
scaler = MinMaxScaler()
scaled_columns = ['Energy (kWh)', 'GHG Savings (kg)', 'Gasoline Savings (gallons)', 'Fee', 'Driver Postal Code', 'EVSE ID', 'System S/N']
df[scaled_columns] = scaler.fit_transform(df[scaled_columns])

# 7. حذف ستون‌هایی که دیگر نیازی ندارند (مثل تاریخ‌های خام)
df.drop(columns=['Start Date', 'End Date'], inplace=True)

# نمایش اطلاعات نهایی
print("اطلاعات نهایی دیتاست پس از پیش‌پردازش:")
print(df.info())
print("\nچند ردیف از دیتاست:")
print(df.head())

# ذخیره دیتاست پردازش شده
df.to_csv("/content/ev_data/EVChargingStationUsage_preprocessed.csv", index=False)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# 1. بارگذاری فایل پیش‌پردازش‌شده
file_path = '/content/ev_data/EVChargingStationUsage_preprocessed.csv'
df = pd.read_csv(file_path)

# 2. پر کردن مقادیر گمشده (در صورت نیاز)

# پر کردن مقادیر گمشده در ستون 'Transaction Date (Pacific Time)' با میانگین تاریخ‌ها
df['Transaction Date (Pacific Time)'] = pd.to_datetime(df['Transaction Date (Pacific Time)'], errors='coerce')
df['Transaction Date (Pacific Time)'].fillna(df['Transaction Date (Pacific Time)'].mean(), inplace=True)

# 3. نرمال‌سازی ستون‌های جغرافیایی (Latitude و Longitude)
scaler = MinMaxScaler()
df[['Latitude', 'Longitude']] = scaler.fit_transform(df[['Latitude', 'Longitude']])

# 4. محاسبه مدت زمان شارژ (ساعت) با استفاده از ویژگی‌های زمانی موجود

# محاسبه زمان شروع و پایان شارژ به صورت یک datetime مجازی
start_time = pd.to_datetime(df[['Start Year', 'Start Month', 'Start Day', 'Start Hour']].astype(str).agg('-'.join, axis=1), format='%Y-%m-%d-%H')
end_time = pd.to_datetime(df[['End Year', 'End Month', 'End Day', 'End Hour']].astype(str).agg('-'.join, axis=1), format='%Y-%m-%d-%H')

# محاسبه مدت زمان شارژ (به ساعت)
df['Charging Duration (hours)'] = (end_time - start_time).dt.total_seconds() / 3600

# مدیریت مقادیر صفر یا منفی در 'Charging Duration (hours)' برای جلوگیری از تقسیم بر صفر
df['Charging Duration (hours)'] = df['Charging Duration (hours)'].replace(0, 0.01)

# 5. محاسبه توان مصرفی (Power) به عنوان نسبت انرژی به مدت زمان شارژ
df['Power (kW)'] = df['Energy (kWh)'] / df['Charging Duration (hours)']

# 6. تبدیل ستون 'User ID' به یک نوع داده یکنواخت (مثلاً str)
df['User ID'] = df['User ID'].astype(str)

# 7. بررسی نهایی
print(df.info())
print(df.head())

# 8. ذخیره دیتای پیش‌پردازش‌شده
df.to_csv('/content/ev_data/EVChargingStationUsage_preprocessed.csv', index=False)
```

```python
import pandas as pd
import numpy as np
import datetime
from meteostat import Point, Daily

# 1. بارگذاری دیتاست شارژ خودروها
file_path = '/content/ev_data/EVChargingStationUsage_preprocessed.csv'
df = pd.read_csv(file_path, low_memory=False)

# 2. تنظیمات مکان و زمان برای جمع‌آوری داده‌های آب و هوا
start = datetime.datetime(2011, 7, 29)  # تاریخ شروع دیتاست شما
end = datetime.datetime(2020, 12, 31)   # تاریخ پایان دیتاست شما
location = Point(37.468319, -122.143936)  # مختصات Palo Alto, CA

# 3. جمع‌آوری داده‌های روزانه دما و بارش با استفاده از Meteostat
weather_data = Daily(location, start, end)
weather_data = weather_data.fetch()

# 4. تبدیل داده‌های آب و هوا به DataFrame
weather_df = weather_data[['tavg', 'prcp']].reset_index()  # tavg: میانگین دما، prcp: بارش
weather_df.columns = ['date', 'temp', 'prcp']  # تغییر نام ستون‌ها به temp و prcp

# 5. ایجاد ستون تاریخ در دیتاست شارژ خودروها از ترکیب Start Year, Start Month و Start Day
# تبدیل مقادیر سال، ماه و روز به رشته و سپس تبدیل به فرمت تاریخ
df['Start Date'] = pd.to_datetime(df['Start Year'].astype(str) + '-' +
                                  df['Start Month'].astype(str) + '-' +
                                  df['Start Day'].astype(str),
                                  errors='coerce')

# بررسی چند ردیف اول از ستون 'Start Date' برای اطمینان از صحت تبدیل
print(df[['Start Date']].head())

# 6. ادغام داده‌های آب و هوا با دیتاست شارژ خودروها بر اساس تاریخ
df['date'] = df['Start Date'].dt.date  # استخراج تاریخ از ستون Start Date
weather_df['date'] = pd.to_datetime(weather_df['date']).dt.date  # تبدیل تاریخ در weather_df به فرمت مناسب
df = pd.merge(df, weather_df, on='date', how='left')

# 7. نرمال‌سازی داده‌های دما و بارش
df['temp_norm'] = df['temp'] / 42  # نرمال‌سازی دما (فرض می‌کنیم حداکثر دما 42 درجه سانتیگراد باشد)
df['prcp_norm'] = df['prcp'] / df['prcp'].max()  # نرمال‌سازی بارش بر اساس حداکثر مقدار

# 8. حذف ستون‌های اضافی
df.drop(columns=['temp', 'prcp', 'date'], inplace=True)

# 9. ذخیره دیتاست نهایی
output_path = '/content/ev_data/EVChargingStationUsage_with_weather.csv'
df.to_csv(output_path, index=False)

# 10. نمایش چند ردیف اول دیتاست نهایی برای بررسی
print(df.head())
```

```python
import pandas as pd
import numpy as np

# 1. بارگذاری دیتاست
file_path = '/content/ev_data/EVChargingStationUsage_with_weather.csv'
df = pd.read_csv(file_path)

# 2. شناسایی مقادیر غیرمنطقی در مدت زمان شارژ و توان مصرفی
print("تعداد مقادیر منفی یا غیرمنطقی در مدت زمان شارژ:")
invalid_duration = df[df['Charging Duration (hours)'] < 0]
print(len(invalid_duration))

print("تعداد مقادیر منفی در توان مصرفی:")
invalid_power = df[df['Power (kW)'] < 0]
print(len(invalid_power))

# 3. مدیریت مقادیر غیرمنطقی
# جایگزین کردن مقادیر منفی در 'Charging Duration (hours)' با میانگین
mean_duration = df['Charging Duration (hours)'].mean()
df.loc[df['Charging Duration (hours)'] < 0, 'Charging Duration (hours)'] = mean_duration

# جایگزین کردن مقادیر منفی در 'Power (kW)' با صفر
df.loc[df['Power (kW)'] < 0, 'Power (kW)'] = 0

# 4. بررسی مجدد توزیع داده‌ها پس از مدیریت مقادیر غیرمنطقی
print("\nتوزیع مدت زمان شارژ پس از مدیریت مقادیر غیرمنطقی:")
print(df['Charging Duration (hours)'].describe())

print("\nتوزیع توان مصرفی پس از مدیریت مقادیر غیرمنطقی:")
print(df['Power (kW)'].describe())

# 5. ذخیره دیتاست نهایی
output_path = '/content/ev_data/EVChargingStationUsage_managed.csv'
df.to_csv(output_path, index=False)
```
