## وارد کردن کتابخانه‌های مورد نیاز


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_olivetti_faces

# بارگیری داده‌ها
X, y = fetch_olivetti_faces(return_X_y=True)
```

##1. محاسبه و نمایش هیستوگرام


```python
def simple_histogram(image):
    # ایجاد یک لیست با 256 عنصر برای نگهداری تعداد پیکسل‌ها در هر شدت رنگ
    hist = [0] * 256
    for pixel in image:
        # تبدیل مقدار پیکسل از [0, 1] به [0, 255]
        hist[int(pixel * 255)] += 1
    return hist

# محاسبه هیستوگرام برای اولین تصویر
first_image = X[0]
histogram = simple_histogram(first_image)

# نمایش هیستوگرام
plt.figure(figsize=(10, 5))
plt.bar(range(256), histogram)
plt.title('Histogram-e Avalin Tasvir')
plt.xlabel('Sheddat-e Pixel')
plt.ylabel('Faravani')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def simple_histogram(image):
    # ایجاد یک لیست با 256 عنصر برای نگهداری تعداد پیکسل‌ها در هر شدت رنگ
    hist = [0] * 256
    for pixel in image:
        # تبدیل مقدار پیکسل از [0, 1] به [0, 255]
        hist[int(pixel * 255)] += 1
    return hist

# محاسبه هیستوگرام برای اولین تصویر
first_image = X[0]
histogram = simple_histogram(first_image)

# تنظیم استایل seaborn برای زیبایی بیشتر
sns.set_style("darkgrid")
plt.figure(figsize=(12, 6))

# رسم هیستوگرام با رنگ‌آمیزی گرادیان
n, bins, patches = plt.hist(range(256), weights=histogram, bins=256, edgecolor='none')

# اعمال رنگ‌آمیزی گرادیان به ستون‌ها
cm = plt.cm.get_cmap('viridis')
for c, p in zip(bins, patches):
    plt.setp(p, 'facecolor', cm(c / 256))

# تنظیم عنوان و برچسب‌ها با فونت فارسی
plt.title('Histogram-e Avalin Tasvir', fontsize=18, fontweight='bold', fontfamily='Arial')
plt.xlabel('Sheddat-e Pixel', fontsize=14)
plt.ylabel('Faravani', fontsize=14)

# اضافه کردن خط میانگین
mean_value = sum(i * v for i, v in enumerate(histogram)) / sum(histogram)
plt.axvline(mean_value, color='r', linestyle='dashed', linewidth=2, label='Miyangin')

# اضافه کردن راهنما
plt.legend(fontsize=12)

# تنظیمات اضافی برای زیبایی بیشتر
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()

# نمایش نمودار
plt.show()
```

##2. قطعه‌بندی ساده تصویر


```python
def simple_segmentation(image, threshold):
    # قطعه‌بندی ساده با استفاده از یک آستانه
    return [1 if pixel > threshold else 0 for pixel in image]

# محاسبه آستانه ساده (میانگین مقادیر پیکسل)
threshold = sum(first_image) / len(first_image)
segmented_image = simple_segmentation(first_image, threshold)

# نمایش تصویر اصلی و قطعه‌بندی شده
plt.figure(figsize=(10, 5))
plt.subplot(121)
plt.imshow(first_image.reshape(64, 64), cmap='gray')
plt.title('Tasvir-e Asli')
plt.subplot(122)
plt.imshow(np.array(segmented_image).reshape(64, 64), cmap='gray')
plt.title('Tasvir-e Ghate-bandi Shode')
plt.show()
```

##3. افزایش وضوح ساده


```python
def simple_super_resolution(image, scale=2):
    original_size = int(np.sqrt(len(image)))
    new_size = original_size * scale
    high_res = [[0 for _ in range(new_size)] for _ in range(new_size)]

    # تکرار هر پیکسل به صورت بلوکی
    for i in range(original_size):
        for j in range(original_size):
            pixel = image[i * original_size + j]
            for x in range(scale):
                for y in range(scale):
                    high_res[i*scale + x][j*scale + y] = pixel

    return high_res

# اعمال افزایش وضوح
high_res_image = simple_super_resolution(first_image)

# نمایش تصویر اصلی و با وضوح بالا
plt.figure(figsize=(10, 5))
plt.subplot(121)
plt.imshow(first_image.reshape(64, 64), cmap='gray')
plt.title('Tasvir-e Asli')
plt.subplot(122)
plt.imshow(high_res_image, cmap='gray')
plt.title('Tasvir ba Vozuh-e Bala')
plt.show()
```

##4. PCA ساده و رگرسیون خطی ساده


```python
def simple_pca(X, n_components):
    # مرکززدایی داده‌ها
    X_centered = X - np.mean(X, axis=0)
    # محاسبه ماتریس کوواریانس
    cov_matrix = np.cov(X_centered.T)
    # محاسبه مقادیر ویژه و بردارهای ویژه
    eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
    # مرتب‌سازی بردارهای ویژه بر اساس مقادیر ویژه
    idx = eigenvalues.argsort()[::-1]
    eigenvectors = eigenvectors[:, idx]
    # انتخاب n_components اول
    components = eigenvectors[:, :n_components]
    return np.dot(X_centered, components)

def simple_linear_regression(X, y):
    # اضافه کردن ستون بایاس
    X_with_bias = np.column_stack((np.ones(X.shape[0]), X))
    # محاسبه ضرایب با استفاده از فرمول تحلیلی
    coeffs = np.linalg.inv(X_with_bias.T @ X_with_bias) @ X_with_bias.T @ y
    return coeffs

def simple_predict(X, coeffs):
    X_with_bias = np.column_stack((np.ones(X.shape[0]), X))
    return X_with_bias @ coeffs

def simple_mse(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

def simple_accuracy(y_true, y_pred):
    return np.sum(np.round(y_pred) == y_true) / len(y_true)

# تقسیم داده‌ها به مجموعه‌های آموزش و آزمون
train_size = int(0.8 * X.shape[0])
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# اعمال PCA ساده
X_pca = simple_pca(X_train, n_components=50)
X_test_pca = simple_pca(X_test, n_components=50)

# آموزش مدل‌ها
coeffs_pca = simple_linear_regression(X_pca, y_train)
coeffs_original = simple_linear_regression(X_train, y_train)

# پیش‌بینی
y_pred_pca = simple_predict(X_test_pca, coeffs_pca)
y_pred_original = simple_predict(X_test, coeffs_original)

# محاسبه MSE و دقت
mse_pca = simple_mse(y_test, y_pred_pca)
mse_original = simple_mse(y_test, y_pred_original)
acc_pca = simple_accuracy(y_test, y_pred_pca)
acc_original = simple_accuracy(y_test, y_pred_original)

print(f'MSE (PCA): {mse_pca:.4f}, Deghat (PCA): {acc_pca:.4f}')
print(f'MSE (Asli): {mse_original:.4f}, Deghat (Asli): {acc_original:.4f}')

# تحلیل ساده نتایج
print("Tahlil-e Sade:")
if mse_pca < mse_original and acc_pca > acc_original:
    print("PCA amalkard-e behtari dasht.")
elif mse_pca > mse_original and acc_pca < acc_original:
    print("Model-e asli amalkard-e behtari dasht.")
else:
    print("Natayej ghati nist va niaz be barresi-e bishtar darad.")
```
