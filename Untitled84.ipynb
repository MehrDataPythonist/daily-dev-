```python
import numpy as np
import pandas as pd
import random
from scipy.spatial.distance import cdist

# تعریف پارامترهای مسئله
N = 200  # تعداد مشتریان
K = 10   # تعداد وسایل نقلیه
Q = 100  # ظرفیت وسایل نقلیه
M = 1000 # جریمه بزرگ

# تولید داده‌های تصادفی برای مشتریان
np.random.seed(42)
locations = np.random.rand(N+1, 2) * 100
demands = np.random.randint(1, 10, size=N)
time_windows = [(random.randint(0, 50), random.randint(50, 100)) for _ in range(N)]
service_times = np.random.randint(1, 10, size=N)

# تخصیص مقادیر v_i و محاسبه α_i
v = np.random.choice([0, 1], size=N, p=[0.7, 0.3])  # 30% مشتریان باید حتماً بازدید شوند
alpha = demands / demands.sum()

# محاسبه ماتریس فاصله
distances = cdist(locations, locations, 'euclidean')
```

 تعریف توابع کمکی

```python
def calculate_distance(locations):
    return cdist(locations, locations, 'euclidean')

def calculate_penalty(solution, time_windows, v, alpha, M):
    penalty = 0
    solution.penalties = [0] * len(time_windows)
    solution.visits = [0] * len(time_windows)
    for route in solution.routes:
        for customer in route:
            t = random.randint(0, 100)  # زمان تصادفی بازدید
            e, l = time_windows[customer]
            if t < e or t > l:
                if v[customer] == 1:
                    penalty += M
                else:
                    penalty += alpha[customer] * (1 - v[customer])
            solution.penalties[customer] = penalty
            solution.visits[customer] = 1 if e <= t <= l else 0
    return penalty
```

تعریف کلاس Solution

```python
class Solution:
    def __init__(self, routes, distances, demands, cost=None):
        self.routes = routes
        self.distances = distances
        self.demands = demands
        self.cost = cost
        self.penalties = [0] * N  # تعداد مشتریان N
        self.visits = [0] * N  # تعداد مشتریان N

    def calculate_cost(self):
        total_cost = 0
        for route in self.routes:
            if route:
                total_cost += sum(self.distances[route[i], route[i+1]] for i in range(len(route) - 1))
        return total_cost

    def is_feasible(self, Q):
        for route in self.routes:
            if sum(self.demands[customer] for customer in route) > Q:
                return False
        return True

    def local_search(self):
        for route in self.routes:
            if len(route) > 1:
                i, j = random.sample(range(len(route)), 2)
                route[i], route[j] = route[j], route[i]
        return Solution(self.routes, self.distances, self.demands, self.calculate_cost())
```

تعریف کلاس ALNS

```python
class ALNS:
    def __init__(self, N, K, Q, M, locations, demands, time_windows, service_times, distances, v, alpha):
        self.N = N
        self.K = K
        self.Q = Q
        self.M = M
        self.locations = locations
        self.demands = demands
        self.time_windows = time_windows
        self.service_times = service_times
        self.distances = distances
        self.v = v
        self.alpha = alpha

    def run(self):
        current_solution = self.create_initial_solution()
        best_solution = current_solution
        best_cost = self.calculate_cost(current_solution)

        for _ in range(1000):  # تعداد تکرارها
            new_solution = self.destroy_and_repair(current_solution)
            new_cost = self.calculate_cost(new_solution)
            if new_cost < best_cost:
                best_solution = new_solution
                best_cost = new_cost
            current_solution = new_solution  # ساده‌سازی، در واقع باید از پذیرش جواب جدید استفاده شود

        return best_solution, best_cost

    def create_initial_solution(self):
        routes = [list(np.random.permutation(self.N)) for _ in range(self.K)]
        return Solution(routes, self.distances, self.demands, self.calculate_cost(Solution(routes, self.distances, self.demands)))

    def destroy_and_repair(self, solution):
        destroyed_routes = solution.routes.copy()
        for route in destroyed_routes:
            if route:
                route.pop(random.randint(0, len(route) - 1))  # حذف تصادفی یک مشتری
        repaired_routes = self.repair(destroyed_routes)
        return Solution(repaired_routes, self.distances, self.demands, self.calculate_cost(Solution(repaired_routes, self.distances, self.demands)))

    def repair(self, routes):
        all_customers = set(range(self.N))
        visited_customers = set(customer for route in routes for customer in route)
        missing_customers = list(all_customers - visited_customers)
        for customer in missing_customers:
            routes[random.randint(0, self.K - 1)].append(customer)  # اضافه کردن به مسیر تصادفی
        return routes

    def calculate_cost(self, solution):
        travel_cost = solution.calculate_cost()
        penalty_cost = calculate_penalty(solution, self.time_windows, self.v, self.alpha, self.M)
        return travel_cost + penalty_cost
```

سلول 5: اجرای الگوریتم و نمایش نتایج

```python
# ایجاد نمونه اولیه از کلاس ALNS
alns = ALNS(N, K, Q, M, locations, demands, time_windows, service_times, distances, v, alpha)

# فراخوانی متد اصلی ALNS برای حل مسئله
best_solution, best_cost = alns.run()

# چاپ و ذخیره نتایج
print("Best Solution: ", best_solution.routes)
print("Best Cost: ", best_cost)

# ایجاد لیست‌های جداگانه برای هر مشتری
route_list = []
penalty_list = best_solution.penalties
visit_list = best_solution.visits

for route in best_solution.routes:
    for customer in route:
        route_list.append(customer)

# ایجاد دیتافریم نتایج
results = pd.DataFrame({
    'مسیر بهینه': route_list,
    'میزان پنالتی': [penalty_list[customer] for customer in route_list],
    'ویزیت در پنجره زمانی': [visit_list[customer] for customer in route_list]
})

# نمایش نتایج
print(results)
```
